# __Recurrent Neural Network (RNN)__

A **Recurrent Neural Network (RNN)** is a special type of Artificial Neural Network (ANN) designed to work with **sequential data**, where the **order of information matters**.

Unlike a traditional ANN, an RNN maintains a **hidden state (memory)** that stores information from previous time steps. This enables the network to learn dependencies within a sequence, making it suitable for tasks involving text, speech, videos, and time-series data.

---

# 📚 __1. Sequential Data__

Sequential data is data in which the **order of elements is important**. Changing the order changes the meaning or context.

### __Examples of Sequential Data__

- 📝 Text
- 🎤 Audio / Speech
- 🎥 Video
- 📈 Time Series Data
- 🧬 DNA Sequences

> **Key Idea:** In sequential data, each element depends on the previous (or surrounding) elements.

### __Example__

Consider the following two sentences:

- **Dog chased cat**
- **Cat chased dog**

Both sentences contain the same words, but the **meaning is completely different** because the **order of the words has changed**.

---

# ⚠️ __2. Limitations of Artificial Neural Networks (ANN)__

Although ANNs perform well on structured data, they are not suitable for sequential data due to the following limitations.

## __1. Order of Data is Ignored__

In an ANN, all input features are processed together without considering their order.

For example,

$$
W_1X_1 + W_2X_2 + W_3X_3
$$

If the order of the inputs changes,

$$
W_3X_3 + W_1X_1 + W_2X_2
$$

the network still performs the same mathematical summation. Thus, an ANN cannot understand that changing the sequence may completely change the meaning of the data.

---

## __2. Fixed Input and Output Size__

An ANN expects a **fixed number of input features** and usually produces a **fixed-size output**.

However, sequential data often has **variable lengths**.

### __Examples__

- Sentence with 5 words
- Sentence with 20 words
- Audio clips of different durations
- Videos containing different numbers of frames

An ANN cannot naturally handle such varying sequence lengths.

---

# ✅ __3. How RNN Solves These Problems__

RNNs overcome the limitations of ANNs by introducing a concept called **recurrence**, where information from previous time steps is remembered and used while processing the current input.

## __1. Preserves Order__

Instead of processing the entire input at once, an RNN processes the sequence **one time step at a time**.

Each new input is processed along with information from previous time steps, allowing the network to learn dependencies within the sequence.

---

## __2. Handles Variable-Length Sequences__

RNNs can process sequences of different lengths because they use **Parameter Sharing**.

### __Parameter Sharing__

The same **weight matrices and biases** are reused at every time step.

Instead of creating a new neural network for every input position, the same network is applied repeatedly across the sequence.

This allows RNNs to process sequences of any length without increasing the number of parameters.

---

# 🔄 __4. Difference Between ANN and RNN__

| Artificial Neural Network (ANN) | Recurrent Neural Network (RNN) |
|---------------------------------|--------------------------------|
| Processes the entire input at once | Processes the input one time step at a time |
| Suitable for non-sequential data | Designed for sequential data |
| Does not remember previous inputs | Maintains information from previous time steps |
| Order of data is ignored | Order of data is preserved |
| Fixed-size input/output | Can handle variable-length sequences |
| Information flows only to the next layer | Information flows to both the next layer and the next time step |

### __Components of an RNN__

- **Input ($x_t$)** → Input at the current time step.
- **RNN Cell** → Processes the current input along with information from previous time steps.
- **Hidden State ($h_t$)** → Stores information (memory) learned from previous inputs.
- **Output ($\hat{y}_t$)** → Prediction produced at the current time step.

### __Key Observation__

Unlike an ANN, the hidden state is **not only forwarded to the next layer but also fed back into the same RNN cell at the next time step**.

This recurrent connection enables the network to retain information from previous inputs while processing a sequence.

> **Note:** The RNN discussed here is called a **Vanilla RNN** (or **Simple RNN**). More advanced variants such as **LSTM** and **GRU** were later developed to overcome some of its limitations.

---

# 🔄 __6. Internal Working of an RNN Cell__

The previous figure represents an RNN as a single block. Internally, each RNN cell receives two inputs:

1. The **current input** ($x_t$)
2. The **previous hidden state** ($h_{t-1}$)

It then computes:

- The new hidden state ($h_t$)
- The output ($\hat{y}_t$)

<div align="center">
    <img src="assets/rnn_cell.png" width="250" alt="Internal Working of an RNN Cell">
    <p><em>Figure: Internal computation of a single RNN cell.</em></p>
</div>

> **Note:** The hidden state ($h_{t-1}$) acts as the **memory** of the network. At every time step, it carries information from previous inputs and combines it with the current input ($x_t$) to produce the new hidden state ($h_t$).

The hidden state is computed as

$$
h_t = f(W_{xh}x_t + W_{hh}h_{t-1} + b)
$$

where

- $x_t$ → Current input
- $h_{t-1}$ → Hidden state from the previous time step
- $h_t$ → Current hidden state
- $W_{xh}$ → Input-to-hidden weight matrix
- $W_{hh}$ → Hidden-to-hidden (recurrent) weight matrix
- $b$ → Bias
- $f(\cdot)$ → Activation function (typically **tanh**)

The activation function introduces **non-linearity**, enabling the network to learn complex patterns from sequential data.

Finally, the output is generated from the hidden state as

$$
\hat{y}_t = g(W_{hy}h_t + b_y)
$$

where

- $W_{hy}$ → Hidden-to-output weight matrix
- $b_y$ → Output bias
- $g(\cdot)$ → Output activation function (depends on the task)

> **Note:** Depending on the application, the output may be generated **at every time step** (e.g., Named Entity Recognition) or **only after processing the complete sequence** (e.g., Sentiment Analysis).

### __Important Points__

- The hidden state acts as the **memory** of the network.
- At every time step, the RNN combines the **current input** with information stored from **previous time steps**.
- The same weight matrices ($W_{xh}$, $W_{hh}$, and $W_{hy}$) are reused across every time step (**Parameter Sharing**).

# 📌 __7. Types of RNN__

Different applications require different input and output sequence structures.

</div>

---

## __1. One-to-One__

- Fixed-size input
- Fixed-size output

This is the traditional ANN architecture rather than a true RNN.

### __Applications__

- House Price Prediction
- Image Classification

---

## __2. One-to-Many__

- Single input
- Variable-length output sequence

### __Applications__

- Image Captioning
- Music Generation

---

## __3. Many-to-One__

- Variable-length input
- Single output

The entire sequence is processed before producing one final prediction.

### __Applications__

- Sentiment Analysis
- Fake News Detection
- Spam Detection
- Movie Review Classification

---

## __4. Many-to-Many (Same Length)__

- Input and output sequences have the same length.

Each input element produces one corresponding output.

### __Applications__

- Named Entity Recognition (NER)
- Part-of-Speech (POS) Tagging

---

## __5. Many-to-Many (Different Length)__

- Input and output sequences have different lengths.

The output sequence does not necessarily have the same number of elements as the input sequence.

### __Applications__

- Machine Translation
- Text Summarization
- Speech Recognition

---

# 📝 __Summary__

- **RNN** is designed for learning from **sequential data**.
- It processes one input at a time while maintaining a **hidden state (memory)**.
- The hidden state enables the network to capture dependencies between previous and current inputs.
- RNN preserves the **order** of information, unlike a traditional ANN.
- The same parameters are reused across every time step (**Parameter Sharing**), allowing variable-length sequences to be processed efficiently.
- RNN architectures are categorized as **One-to-One**, **One-to-Many**, **Many-to-One**, and **Many-to-Many**, depending on the input-output sequence structure.
- Although **Vanilla RNNs** work well for short sequences, they struggle with **long-term dependencies** due to the **Vanishing Gradient Problem**. This limitation led to the development of more advanced architectures such as **LSTM** and **GRU**.
