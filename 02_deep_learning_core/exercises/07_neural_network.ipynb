# 🧠 __Neural Networks__

A **Neural Network (NN)** is a machine learning model inspired by the way the **human brain** processes information.

It consists of interconnected **neurons (nodes)** that learn patterns from data by adjusting their **weights** during training.

> **Goal:** Learn patterns from data and make accurate predictions.

---

# 🏗️ __Types of Neural Networks__

## __1. Artificial Neural Network (ANN)__

- Best suited for **Structured / Tabular Data**
- Consists of:
  - Input Layer
  - Hidden Layer(s)
  - Output Layer

### __Applications__
- House Price Prediction
- Customer Churn Prediction
- Credit Risk Analysis
- Sales Forecasting

---

## __2. Convolutional Neural Network (CNN)__

- Best suited for **Image and Video Data**
- Automatically extracts features like edges, textures, and shapes.

### __Applications__
- Image Classification
- Object Detection
- Face Recognition
- Medical Image Analysis

---

## __3. Recurrent Neural Network (RNN)__

- Best suited for **Sequential Data**
- Remembers previous information using hidden states.

### __Applications__
- Natural Language Processing (NLP)
- Speech Recognition
- Time Series Forecasting
- Language Translation

---

# 🧠 __Basic Structure of an ANN__

```text
           Input Layer          Hidden Layer(s)          Output Layer

          x₁  ●  ───────────────►  ●
          x₂  ●  ───────────────►  ●  ───────────────►  ●
          x₃  ●  ───────────────►  ●
```

Each connection between neurons has a **Weight**.

Every neuron performs the following computation:

$$
z = w_1x_1 + w_2x_2 + \cdots + w_nx_n + b
$$

$$
a = f(z)
$$

Where

- **x** → Input Features
- **w** → Weights
- **b** → Bias
- **z** → Weighted Sum
- **f()** → Activation Function
- **a** → Output

---

# 📈 __Why Do We Need Derivatives?__

A Neural Network learns by **updating its weights**.

To know **how much each weight should change**, we calculate **Derivatives**.

A derivative tells us:

> **How much does the output change when the input changes slightly?**

---

# 📚 __Derivatives__

A **Derivative** measures the **rate of change** of one variable with respect to another.

Mathematically,

$$
\frac{dy}{dx}
$$

### Meaning

- **dy** → Small change in output
- **dx** → Small change in input

So,

$$
\frac{dy}{dx}
$$

means

> **Change in output with respect to change in input**

---

## __Example__

Given,

$$
y = x^2
$$

Differentiate both sides with respect to \(x\):

$$
\frac{dy}{dx}
=
\frac{d}{dx}(x^2)
=
2x
$$

Thus, the derivative of the function is

$$
\boxed{\frac{dy}{dx}=2x}
$$

Now, evaluate the derivative at \(x=3\):

$$
\left.\frac{dy}{dx}\right|_{x=3}
=
2(3)
=
6
$$

Hence,

$$
\boxed{\left.\frac{dy}{dx}\right|_{x=3}=6}
$$

### __Interpretation__

At the point \(x=3\), the slope of the curve \(y=x^2\) is **6**.

This means that for a very small change in \(x\) around \(x=3\), the value of \(y\) changes approximately **6 times** that amount.

# 📚 __Partial Derivatives__

When a function has **more than one input variable**, we calculate the derivative with respect to **one variable while keeping all others constant**.

Instead of using

$$
\frac{dy}{dx}
$$

we use

$$
\frac{\partial y}{\partial x}
$$

---

## __Example__

Suppose,

$$
y=3x+2z
$$

Then,

$$
\frac{\partial y}{\partial x}=3
$$

because **z** is treated as a constant.

Similarly,

$$
\frac{\partial y}{\partial z}=2
$$

because **x** is treated as a constant.

---

## __Why Partial Derivatives?__

Neural Networks contain **thousands or even millions of weights**.

To update each weight independently, we compute

$$
\frac{\partial J}{\partial W_1},
\quad
\frac{\partial J}{\partial W_2},
\quad
...
$$

> **Note**
>
> Since a Neural Network contains many parameters (weights and biases), we use **Partial Derivatives** instead of ordinary derivatives to compute how each parameter affects the loss independently.

# 🔗 __Chain Rule__

The **Chain Rule** is used to calculate the derivative of a **composite function**, i.e., a function that is made up of multiple smaller functions.

Instead of differentiating the entire expression at once, we:

1. Break the function into simpler intermediate variables.
2. Compute the derivative of each intermediate variable.
3. Multiply the derivatives along the path.

This is the fundamental idea behind **Backpropagation** in Neural Networks.

---

## __Example__

Consider the function

$$
y = 3(a + bc)
$$

### __Step 1: Break the function into smaller parts__

Define intermediate variables:

$$
u = bc
$$

$$
v = a + u
$$

$$
y = 3v
$$

The computational graph becomes

```text
a ───────────────► (+) ─────────► (×3) ─────────► y
                   ▲
                   │
b ─────► (×) ──────┘
         ▲
         │
c ───────┘

u = bc
v = a + u
y = 3v
```

---

### __Step 2: Compute the local derivatives__

From

$$
y = 3v
$$

$$
\frac{dy}{dv}=3
$$

From

$$
v = a + u
$$

$$
\frac{dv}{da}=1
$$

$$
\frac{dv}{du}=1
$$

From

$$
u = bc
$$

$$
\frac{du}{db}=c
$$

$$
\frac{du}{dc}=b
$$

---

### __Step 3: Apply the Chain Rule__

#### __Derivative with respect to \(a\)__

There is only one path:

$$
a \rightarrow v \rightarrow y
$$

Therefore,

$$
\frac{dy}{da}
=
\frac{dy}{dv}
\times
\frac{dv}{da}
=
3 \times 1
=
3
$$

---

#### __Derivative with respect to \(b\)__

The path is

$$
b \rightarrow u \rightarrow v \rightarrow y
$$

Hence,

$$
\frac{dy}{db}
=
\frac{dy}{dv}
\times
\frac{dv}{du}
\times
\frac{du}{db}
$$

Substituting the values,

$$
\frac{dy}{db}
=
3 \times 1 \times c
=
3c
$$

---

#### __Derivative with respect to \(c\)__

The path is

$$
c \rightarrow u \rightarrow v \rightarrow y
$$

Hence,

$$
\frac{dy}{dc}
=
\frac{dy}{dv}
\times
\frac{dv}{du}
\times
\frac{du}{dc}
$$

Substituting the values,

$$
\frac{dy}{dc}
=
3 \times 1 \times b
=
3b
$$

---

## 📌 __Final Results__

$$
\boxed{
\frac{dy}{da}=3
}
$$

$$
\boxed{
\frac{dy}{db}=3c
}
$$

$$
\boxed{
\frac{dy}{dc}=3b
}
$$

---

## 💡 __Why is the Chain Rule Important?__

A Neural Network is simply a **composition of many functions**.

```text
Input
   │
   ▼
Weighted Sum
   │
   ▼
Activation Function
   │
   ▼
Loss Function
```

To calculate how each **weight** affects the final **loss**, we repeatedly multiply local derivatives using the **Chain Rule**.

This entire process is called **Backpropagation**, which is the core learning algorithm used to train Neural Networks.

> **Note**
>
> Modern Deep Learning frameworks such as **PyTorch** and **TensorFlow** automatically compute these derivatives using **Automatic Differentiation (Autograd)**, which internally applies the **Chain Rule**.

# ⚙️ __Weight Updation__

After computing the gradients using **Backpropagation**, the neural network updates its weights to reduce the loss.

This optimization process is called **Gradient Descent**.

The update rule is

$$
W_{\text{new}}
=
W_{\text{old}}
-
\eta
\frac{\partial J}{\partial W}
$$

Where

- **W** → Weight
- **J** → Cost (Loss) Function
- **η (Eta)** → Learning Rate
- **∂J/∂W** → Gradient

---

## __Gradient ( ∂J / ∂W )__

Gradient tells us

- Which direction to move
- How sensitive the loss is to the weight

If the gradient is

- Positive → Decrease the weight
- Negative → Increase the weight
- Zero → Weight is close to the optimum value

---

## __Learning Rate (η)__

Learning Rate controls **how big a step** we take while updating the weights.

### __Small Learning Rate__

- Slow learning
- Stable convergence

### __Large Learning Rate__

- Faster learning
- May overshoot the minimum
- Can fail to converge

# 🧠 __How a Neuron Works__

A **Neuron** is the fundamental building block of a Neural Network.

Each neuron receives **input features**, performs mathematical computations, and produces an output that is passed to the next layer.

Every neuron performs **two main functions**:

1. **Summation Function (Linear Transformation)**
2. **Activation Function (Non-linear Transformation)**

---

# ➕ __1. Summation Function__

The neuron first computes a **weighted sum** of all input features and then adds a **bias**.

The mathematical equation is

$$
Z = W_1X_1 + W_2X_2 + \cdots + W_nX_n + B
$$

Where

- **X** → Input Features
- **W** → Weights
- **B** → Bias
- **Z** → Weighted Sum (Linear Output)

The bias helps shift the output and gives the neuron more flexibility in learning.

---

# ⚡ __2. Activation Function__

The output of the summation function is passed through an **Activation Function**.

$$
A = f(Z)
$$

Where

- **Z** → Output of Summation Function
- **f()** → Activation Function
- **A** → Final Output of the Neuron

The activation function introduces **non-linearity**, allowing the neural network to learn complex patterns.

Without an activation function, a neural network behaves like a simple linear model regardless of how many layers it has.

# 🎯 __Which Activation Function to Use?__

The choice of activation function depends on the type of problem and the layer in which it is used.

---

## __Output Layer__

### __Classification Problems__

#### __1. Binary Classification__

Use **Sigmoid** Activation.

**Examples:**

- Spam Detection
- Disease Prediction
- Fake News Detection

---

#### __2. Multi-class Classification__

Use **Softmax** Activation.

**Examples:**

- Digit Recognition
- Animal Classification
- Flower Classification

---

### __3. Regression Problems__

Use **Linear Activation (No Activation Function)**

**Examples:**

- House Price Prediction
- Sales Prediction
- Temperature Prediction

---

## __Hidden Layer__

Hidden layers generally use **Non-linear Activation Functions**.

The most commonly used activation is

- **ReLU (Rectified Linear Unit)**

Other activation functions include

- Sigmoid
- Tanh
- Leaky ReLU
- ELU

> **Note**
>
> ReLU is the most popular activation function for hidden layers because it is simple, computationally efficient, and helps deep neural networks train faster.

---

# 🔄 __How Training in a Neural Network Works__

Training is the process of adjusting the **weights** and **biases** so that the neural network makes accurate predictions.

---

## __Step 1: Input Features__

Input features are passed to the **Input Layer**.

---

## __Step 2: Weight Initialization__

Assign **random values** to all weights and biases.

This is called **Weight Initialization**.

Initially, the network does not know anything about the data.

---

## __Step 3: Summation Function__

Each neuron computes

$$
Z = W_1X_1 + W_2X_2 + \cdots + W_nX_n + B
$$

---

## __Step 4: Activation Function__

The weighted sum is passed through an activation function.

$$
A = f(Z)
$$

This introduces **non-linearity** into the network.

---

## __Step 5: Pass Output to Next Layer__

The output of one layer becomes the input to the next layer.

This process continues until the output layer is reached.

---

## __Step 6: Final Prediction__

The output layer produces the final prediction.

---

# ➡️ __Forward Propagation__

Steps **1 to 6** together are called

> **Forward Propagation**

During forward propagation, data moves from the **Input Layer** to the **Output Layer** to generate predictions.

---

## __Step 7: Compute Loss__

Compare the **Predicted Output** with the **Actual Output**.

The difference between them is called the **Loss**.

A smaller loss indicates better predictions.

---

## __Step 8: Update Weights and Biases__

Using **Backpropagation**, calculate the gradients of the loss with respect to each weight and bias.

Then update them to reduce the loss.

---

# ⬅️ __Backward Propagation__

Step **8** is called

> **Backward Propagation**

During backpropagation, gradients are calculated using the **Chain Rule**, and weights are updated using **Gradient Descent**.

---

## __Step 9: Repeat__

Repeat the following steps until the loss becomes sufficiently small:

- Forward Propagation
- Loss Calculation
- Backpropagation
- Weight Update

This iterative process allows the neural network to learn from the training data.

---

# ⚙️ __How Weights and Biases are Updated__

The update equations are

$$
W_{\text{new}}
=
W_{\text{old}}
-
\eta
\frac{\partial J}{\partial W}
$$

$$
b_{\text{new}}
=
b_{\text{old}}
-
\eta
\frac{\partial J}{\partial b}
$$

Where

- **W** → Weight
- **b** → Bias
- **J** → Loss Function
- **η (Eta)** → Learning Rate
- **∂J/∂W** → Gradient of Loss with respect to Weight
- **∂J/∂b** → Gradient of Loss with respect to Bias

---

# 📉 __Gradient Descent__

The optimization algorithm used to update weights is called **Gradient Descent**.

It consists of two terms:

## __Gradient__

The **Gradient** is the derivative of the loss function with respect to the weights.

$$
\frac{\partial J}{\partial W}
$$

It tells us

- Which direction to move.
- How much the loss changes when the weight changes.

---

## __Descent__

Descent means moving **downhill** toward the minimum value of the loss function.

The objective is to find the weights that produce the **minimum loss**.

---

## __Relationship__

```text
Slope
   │
   ▼
Derivative
   │
   ▼
Gradient
   │
   ▼
Gradient Descent
```

---

# 🔄 __Complete Neural Network Training Pipeline__

```text
Input Features
       │
       ▼
Weight Initialization
       │
       ▼
Summation Function
       │
       ▼
Activation Function
       │
       ▼
Forward Propagation
       │
       ▼
Prediction
       │
       ▼
Loss Calculation
       │
       ▼
Backward Propagation
       │
       ▼
Gradient Calculation
       │
       ▼
Update Weights & Biases
       │
       ▼
Repeat Until Loss is Minimum
```

# 📌 __Key Takeaways__

- A neuron performs two operations: **Summation** and **Activation**.
- The summation function computes the weighted sum of inputs and bias.
- Activation functions introduce **non-linearity**, enabling the network to learn complex patterns.
- ReLU is the most commonly used activation function in hidden layers.
- Sigmoid is used for **Binary Classification**.
- Softmax is used for **Multi-class Classification**.
- Linear activation is used for **Regression**.
- A layer is a collection of neurons, and multiple layers form a neural network.
- Forward Propagation computes predictions.
- Backpropagation computes gradients using the **Chain Rule**.
- Gradient Descent updates weights and biases to minimize the loss.
- Training continues until the loss reaches a minimum or an acceptable value.

# 🚀 __Improving Neural Networks__

Training a Neural Network is not only about designing the architecture but also about choosing the right optimization techniques, hyperparameters, and methods to improve its generalization performance.

---

# 📈 __Variants of Gradient Descent__

Gradient Descent is an optimization algorithm used to update the **weights** and **biases** of a neural network to minimize the loss function.

There are three main variants of Gradient Descent based on how much data is used to calculate the gradient.

---

## __1. Batch Gradient Descent__

Batch Gradient Descent uses the **entire training dataset** to compute the gradient before updating the weights.

### __Characteristics__

- Uses all training samples.
- One weight update after processing the complete dataset.
- Stable and accurate gradient.
- Slow for large datasets.
- Requires high memory.

---

## __2. Stochastic Gradient Descent (SGD)__

Stochastic Gradient Descent updates the weights using **only one training sample** at a time.

### __Characteristics__

- Uses one sample for each weight update.
- Very fast updates.
- Can escape local minima because of noisy updates.
- Training is less stable.

---

## __3. Mini Batch Gradient Descent__

Mini Batch Gradient Descent is the most commonly used approach.

Instead of using one sample or the entire dataset, it uses a **small batch of samples** to compute the gradient.

### __Characteristics__

- Faster than Batch Gradient Descent.
- More stable than Stochastic Gradient Descent.
- Efficient on modern GPUs.
- Most deep learning libraries use Mini Batch Gradient Descent by default.

---

## 📌 __Relationship Between Batch Size and Gradient Descent__

| Batch Size | Gradient Descent Type |
|------------|-----------------------|
| Batch Size = 1 | Stochastic Gradient Descent (SGD) |
| Batch Size = Total Dataset | Batch Gradient Descent |
| 1 < Batch Size < Total Dataset | Mini Batch Gradient Descent |

> **Note**
>
> Mini Batch Gradient Descent is the most commonly used approach in Deep Learning because it balances speed and stability.

---

# ⚙️ __Optimizers__

An **Optimizer** is an algorithm that updates the weights and biases using the gradients computed during backpropagation.

Different optimizers improve the speed and stability of training.

Some popular optimizers are

- Gradient Descent with Momentum
- AdaGrad
- RMSProp
- Adam

---

## __1. Gradient Descent with Momentum__

Momentum helps the optimizer move faster in the correct direction by remembering previous updates.

### __Advantages__

- Faster convergence.
- Reduces oscillations.
- Helps escape shallow local minima.

---

## __2. AdaGrad__

AdaGrad adapts the learning rate for each parameter individually.

### __Advantages__

- Good for sparse data.
- Larger updates for infrequent parameters.
- Smaller updates for frequent parameters.

### __Limitation__

The learning rate keeps decreasing during training, which may slow learning.

---

## __3. RMSProp__

RMSProp improves AdaGrad by preventing the learning rate from becoming too small.

### __Advantages__

- Faster convergence.
- Works well for deep neural networks.
- Commonly used for recurrent neural networks.

---

## __4. Adam (Adaptive Moment Estimation)__

Adam combines the advantages of **Momentum** and **RMSProp**.

### __Advantages__

- Fast convergence.
- Adaptive learning rates.
- Stable training.
- Works well for most deep learning problems.

> **Default Recommendation**
>
> ✅ Use **Adam** optimizer unless you have a specific reason to choose another optimizer.

# 📚 __Important Terms__

---

## __1. Epoch__

An **Epoch** is one complete pass of the entire training dataset through the neural network.

```text
1 Epoch
=
1 Forward Pass
+
1 Backward Pass
on the complete training dataset
```

---

## __2. Sample__

A **Sample** is a single data point in the dataset.

**Example**

```text
Age  Salary  Purchased
25   40000      Yes
```

This single row is one sample.

---

## __3. Batch Size__

Batch Size is the **number of samples processed before updating the weights**.

**Examples**

- Batch Size = 1 → SGD
- Batch Size = 32
- Batch Size = 64
- Batch Size = 128

---

## __4. Iteration__

An **Iteration** is one weight update.

The number of iterations required to complete one epoch is

$$
\text{Iterations}
=
\frac{\text{Number of Samples}}{\text{Batch Size}}
$$

**Example**

```text
Dataset Size = 1000
Batch Size = 100

Iterations = 1000 / 100 = 10
```

---

## __5. Learning Rate__

The **Learning Rate (η)** is a hyperparameter that controls how much the weights are updated during training.

### __Small Learning Rate__

- Slow learning.
- Stable convergence.

### __Large Learning Rate__

- Faster learning.
- May overshoot the minimum loss.

---

## __6. Parameters vs Hyperparameters__

### __Parameters__

Parameters are learned automatically during training.

**Examples**

- Weights
- Biases

---

### __Hyperparameters__

Hyperparameters are chosen **before training**.

**Examples**

- Learning Rate
- Batch Size
- Number of Epochs
- Number of Hidden Layers
- Number of Neurons
- Optimizer

---

## __7. Cost Function vs Loss Function__

### __Loss Function__

The **Loss Function** measures the error for a **single training sample**.

```text
One Sample
      │
      ▼
Loss
```

---

### __Cost Function__

The **Cost Function** is the average loss over the **entire training dataset**.

```text
All Samples
      │
      ▼
Average Loss
      │
      ▼
Cost
```

> **Note**
>
> In practice, the terms *Loss Function* and *Cost Function* are often used interchangeably, although they have different meanings.

# 🚀 __Improving Neural Networks__

---

# 📏 __1. Scale the Input Features__

Features in a dataset may have different ranges.

**Example**

```text
Age       : 25
Salary    : 500000
```

Since **Salary** has much larger values, the model may give it more importance during training.

To avoid this, we scale all features to a similar range.

Common scaling techniques

- MinMaxScaler
- StandardScaler

### __Benefits__

- Faster convergence.
- Stable gradient updates.
- Better model performance.
- Prevents features with larger values from dominating learning.

---

# 📉 __2. Underfitting & Overfitting__

A good neural network should perform well on both **training** and **testing** data.

---

# ❌ __Underfitting__

Underfitting occurs when the model fails to learn the underlying patterns in the data.

### __Characteristics__

- Poor performance on training data.
- Poor performance on testing data.

```text
Training Accuracy  → Low
Testing Accuracy   → Low
```

---

## __How to Handle Underfitting__

- Train for more epochs.
- Collect more training data.
- Increase model complexity.
  - More hidden layers.
  - More neurons.
- Remove noise from the data.
- Perform hyperparameter tuning.

---

# ⚠️ __Overfitting__

Overfitting occurs when the model memorizes the training data instead of learning general patterns.

### __Characteristics__

- High training accuracy.
- Low testing accuracy.

```text
Training Accuracy  → High
Testing Accuracy   → Low
```

---

## __How to Handle Overfitting__

### __Add More Diverse Data__

A larger and more diverse dataset helps the model generalize better.

---

### __Regularization__

Regularization discourages the model from becoming overly complex.

Common techniques

- L1 Regularization
- L2 Regularization

---

### __Dropout__

Dropout randomly disables some neurons during training.

Benefits

- Prevents co-adaptation of neurons.
- Reduces overfitting.
- Improves generalization.

---

### __Early Stopping__

Early Stopping monitors the validation performance during training.

If the validation loss stops improving, training is stopped automatically.

This prevents the model from overfitting.

# 📌 __Key Takeaways__

- Gradient Descent has three variants: **Batch**, **Stochastic (SGD)**, and **Mini Batch**.
- Mini Batch Gradient Descent is the most commonly used variant in deep learning.
- **Adam** is the recommended optimizer for most problems.
- One **Epoch** is one complete forward and backward pass over the entire dataset.
- **Iteration = Dataset Size / Batch Size**.
- Parameters are learned during training, while Hyperparameters are set before training.
- A **Loss Function** measures the error for a single sample, whereas a **Cost Function** is the average loss over the entire dataset.
- Use **MSE** or **MAE** for regression problems.
- Use **Binary Cross Entropy** for binary classification and **Categorical Cross Entropy** for multi-class classification.
- Scaling input features with **MinMaxScaler** or **StandardScaler** improves training stability and convergence.
- **Underfitting** means the model performs poorly on both training and testing data, while **Overfitting** means it performs well on training data but poorly on unseen data.
- Common techniques to reduce overfitting include **L1/L2 Regularization**, **Dropout**, **Early Stopping**, and using more diverse training data.
