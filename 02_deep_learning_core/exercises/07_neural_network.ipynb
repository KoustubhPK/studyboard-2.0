# 🧠 __Neural Networks__

A **Neural Network (NN)** is a machine learning model inspired by the way the **human brain** processes information.

It consists of interconnected **neurons (nodes)** that learn patterns from data by adjusting their **weights** during training.

> **Goal:** Learn patterns from data and make accurate predictions.

---

# 🏗️ __Types of Neural Networks__

## __1. Artificial Neural Network (ANN)__

- Best suited for **Structured / Tabular Data**
- Consists of:
  - Input Layer
  - Hidden Layer(s)
  - Output Layer

### __Applications__
- House Price Prediction
- Customer Churn Prediction
- Credit Risk Analysis
- Sales Forecasting

---

## __2. Convolutional Neural Network (CNN)__

- Best suited for **Image and Video Data**
- Automatically extracts features like edges, textures, and shapes.

### __Applications__
- Image Classification
- Object Detection
- Face Recognition
- Medical Image Analysis

---

## __3. Recurrent Neural Network (RNN)__

- Best suited for **Sequential Data**
- Remembers previous information using hidden states.

### __Applications__
- Natural Language Processing (NLP)
- Speech Recognition
- Time Series Forecasting
- Language Translation

---

# 🧠 __Basic Structure of an ANN__

```text
           Input Layer          Hidden Layer(s)          Output Layer

          x₁  ●  ───────────────►  ●
          x₂  ●  ───────────────►  ●  ───────────────►  ●
          x₃  ●  ───────────────►  ●
```

Each connection between neurons has a **Weight**.

Every neuron performs the following computation:

$$
z = w_1x_1 + w_2x_2 + \cdots + w_nx_n + b
$$

$$
a = f(z)
$$

Where

- **x** → Input Features
- **w** → Weights
- **b** → Bias
- **z** → Weighted Sum
- **f()** → Activation Function
- **a** → Output

---

# 📈 __Why Do We Need Derivatives?__

A Neural Network learns by **updating its weights**.

To know **how much each weight should change**, we calculate **Derivatives**.

A derivative tells us:

> **How much does the output change when the input changes slightly?**

---

# 📚 __Derivatives__

A **Derivative** measures the **rate of change** of one variable with respect to another.

Mathematically,

$$
\frac{dy}{dx}
$$

### Meaning

- **dy** → Small change in output
- **dx** → Small change in input

So,

$$
\frac{dy}{dx}
$$

means

> **Change in output with respect to change in input**

---

## __Example__

Given,

$$
y = x^2
$$

Differentiate both sides with respect to \(x\):

$$
\frac{dy}{dx}
=
\frac{d}{dx}(x^2)
=
2x
$$

Thus, the derivative of the function is

$$
\boxed{\frac{dy}{dx}=2x}
$$

Now, evaluate the derivative at \(x=3\):

$$
\left.\frac{dy}{dx}\right|_{x=3}
=
2(3)
=
6
$$

Hence,

$$
\boxed{\left.\frac{dy}{dx}\right|_{x=3}=6}
$$

### __Interpretation__

At the point \(x=3\), the slope of the curve \(y=x^2\) is **6**.

This means that for a very small change in \(x\) around \(x=3\), the value of \(y\) changes approximately **6 times** that amount.

# 📚 __Partial Derivatives__

When a function has **more than one input variable**, we calculate the derivative with respect to **one variable while keeping all others constant**.

Instead of using

$$
\frac{dy}{dx}
$$

we use

$$
\frac{\partial y}{\partial x}
$$

---

## __Example__

Suppose,

$$
y=3x+2z
$$

Then,

$$
\frac{\partial y}{\partial x}=3
$$

because **z** is treated as a constant.

Similarly,

$$
\frac{\partial y}{\partial z}=2
$$

because **x** is treated as a constant.

---

## __Why Partial Derivatives?__

Neural Networks contain **thousands or even millions of weights**.

To update each weight independently, we compute

$$
\frac{\partial J}{\partial W_1},
\quad
\frac{\partial J}{\partial W_2},
\quad
...
$$

> **Note**
>
> Since a Neural Network contains many parameters (weights and biases), we use **Partial Derivatives** instead of ordinary derivatives to compute how each parameter affects the loss independently.

# 🔗 __Chain Rule__

The **Chain Rule** is used to calculate the derivative of a **composite function**, i.e., a function that is made up of multiple smaller functions.

Instead of differentiating the entire expression at once, we:

1. Break the function into simpler intermediate variables.
2. Compute the derivative of each intermediate variable.
3. Multiply the derivatives along the path.

This is the fundamental idea behind **Backpropagation** in Neural Networks.

---

## __Example__

Consider the function

$$
y = 3(a + bc)
$$

### __Step 1: Break the function into smaller parts__

Define intermediate variables:

$$
u = bc
$$

$$
v = a + u
$$

$$
y = 3v
$$

The computational graph becomes

```text
a ───────────────► (+) ─────────► (×3) ─────────► y
                   ▲
                   │
b ─────► (×) ──────┘
         ▲
         │
c ───────┘

u = bc
v = a + u
y = 3v
```

---

### __Step 2: Compute the local derivatives__

From

$$
y = 3v
$$

$$
\frac{dy}{dv}=3
$$

From

$$
v = a + u
$$

$$
\frac{dv}{da}=1
$$

$$
\frac{dv}{du}=1
$$

From

$$
u = bc
$$

$$
\frac{du}{db}=c
$$

$$
\frac{du}{dc}=b
$$

---

### __Step 3: Apply the Chain Rule__

#### __Derivative with respect to \(a\)__

There is only one path:

$$
a \rightarrow v \rightarrow y
$$

Therefore,

$$
\frac{dy}{da}
=
\frac{dy}{dv}
\times
\frac{dv}{da}
=
3 \times 1
=
3
$$

---

#### __Derivative with respect to \(b\)__

The path is

$$
b \rightarrow u \rightarrow v \rightarrow y
$$

Hence,

$$
\frac{dy}{db}
=
\frac{dy}{dv}
\times
\frac{dv}{du}
\times
\frac{du}{db}
$$

Substituting the values,

$$
\frac{dy}{db}
=
3 \times 1 \times c
=
3c
$$

---

#### __Derivative with respect to \(c\)__

The path is

$$
c \rightarrow u \rightarrow v \rightarrow y
$$

Hence,

$$
\frac{dy}{dc}
=
\frac{dy}{dv}
\times
\frac{dv}{du}
\times
\frac{du}{dc}
$$

Substituting the values,

$$
\frac{dy}{dc}
=
3 \times 1 \times b
=
3b
$$

---

## 📌 __Final Results__

$$
\boxed{
\frac{dy}{da}=3
}
$$

$$
\boxed{
\frac{dy}{db}=3c
}
$$

$$
\boxed{
\frac{dy}{dc}=3b
}
$$

---

## 💡 __Why is the Chain Rule Important?__

A Neural Network is simply a **composition of many functions**.

```text
Input
   │
   ▼
Weighted Sum
   │
   ▼
Activation Function
   │
   ▼
Loss Function
```

To calculate how each **weight** affects the final **loss**, we repeatedly multiply local derivatives using the **Chain Rule**.

This entire process is called **Backpropagation**, which is the core learning algorithm used to train Neural Networks.

> **Note**
>
> Modern Deep Learning frameworks such as **PyTorch** and **TensorFlow** automatically compute these derivatives using **Automatic Differentiation (Autograd)**, which internally applies the **Chain Rule**.

# ⚙️ __Weight Updation__

After computing the gradients using **Backpropagation**, the neural network updates its weights to reduce the loss.

This optimization process is called **Gradient Descent**.

The update rule is

$$
W_{\text{new}}
=
W_{\text{old}}
-
\eta
\frac{\partial J}{\partial W}
$$

Where

- **W** → Weight
- **J** → Cost (Loss) Function
- **η (Eta)** → Learning Rate
- **∂J/∂W** → Gradient

---

## __Gradient ( ∂J / ∂W )__

Gradient tells us

- Which direction to move
- How sensitive the loss is to the weight

If the gradient is

- Positive → Decrease the weight
- Negative → Increase the weight
- Zero → Weight is close to the optimum value

---

## __Learning Rate (η)__

Learning Rate controls **how big a step** we take while updating the weights.

### __Small Learning Rate__

- Slow learning
- Stable convergence

### __Large Learning Rate__

- Faster learning
- May overshoot the minimum
- Can fail to converge

# 🔄 __Neural Network Learning Process__

```text
Initialize Weights
        │
        ▼
Forward Propagation
        │
        ▼
Prediction
        │
        ▼
Compute Loss
        │
        ▼
Backpropagation
        │
        ▼
Calculate Gradients
        │
        ▼
Update Weights
        │
        ▼
Repeat Until Loss is Minimum
```

# 📌 __Key Takeaways__

- Neural Networks learn by adjusting **weights**.
- Derivatives measure how output changes with input.
- Partial derivatives are used when there are multiple input variables.
- The Chain Rule calculates gradients through multiple layers.
- Backpropagation uses the Chain Rule to compute gradients.
- Gradient Descent updates weights to reduce the loss.
- Learning Rate controls the speed of learning.

# 🧠 __How a Neuron Works__

A **Neuron** is the fundamental building block of a Neural Network.

Each neuron receives **input features**, performs mathematical computations, and produces an output that is passed to the next layer.

Every neuron performs **two main functions**:

1. **Summation Function (Linear Transformation)**
2. **Activation Function (Non-linear Transformation)**

---

# ➕ __1. Summation Function__

The neuron first computes a **weighted sum** of all input features and then adds a **bias**.

The mathematical equation is

$$
Z = W_1X_1 + W_2X_2 + \cdots + W_nX_n + B
$$

Where

- **X** → Input Features
- **W** → Weights
- **B** → Bias
- **Z** → Weighted Sum (Linear Output)

The bias helps shift the output and gives the neuron more flexibility in learning.

---

# ⚡ __2. Activation Function__

The output of the summation function is passed through an **Activation Function**.

$$
A = f(Z)
$$

Where

- **Z** → Output of Summation Function
- **f()** → Activation Function
- **A** → Final Output of the Neuron

The activation function introduces **non-linearity**, allowing the neural network to learn complex patterns.

Without an activation function, a neural network behaves like a simple linear model regardless of how many layers it has.

# 🎯 __Which Activation Function to Use?__

The choice of activation function depends on the type of problem and the layer in which it is used.

---

## __Output Layer__

### __Classification Problems__

#### __1. Binary Classification__

Use **Sigmoid** Activation.

**Examples:**

- Spam Detection
- Disease Prediction
- Fake News Detection

---

#### __2. Multi-class Classification__

Use **Softmax** Activation.

**Examples:**

- Digit Recognition
- Animal Classification
- Flower Classification

---

### __3. Regression Problems__

Use **Linear Activation (No Activation Function)**

**Examples:**

- House Price Prediction
- Sales Prediction
- Temperature Prediction

---

## __Hidden Layer__

Hidden layers generally use **Non-linear Activation Functions**.

The most commonly used activation is

- **ReLU (Rectified Linear Unit)**

Other activation functions include

- Sigmoid
- Tanh
- Leaky ReLU
- ELU

> **Note**
>
> ReLU is the most popular activation function for hidden layers because it is simple, computationally efficient, and helps deep neural networks train faster.

---

A **Layer** is a collection of neurons.

When multiple neurons are stacked together, they form a **Layer**.

When multiple layers are connected together, they form a **Neural Network**.

# 📚 __Types of Layers__

## __1. Input Layer__

- First layer of the neural network.
- Receives the input features from the dataset.
- Does not perform any computation.

Examples of input features:

- Age
- Salary
- Area
- Number of Bedrooms

---

## __2. Hidden Layer__

- Performs most of the computations.
- Calculates weighted sums.
- Applies activation functions.
- Learns complex patterns from the input data.

A neural network can have **one or multiple hidden layers**.

---

## __3. Output Layer__

- Produces the final prediction.
- Uses an activation function based on the problem type.

Examples:

- Binary Classification → Sigmoid
- Multi-class Classification → Softmax
- Regression → Linear

---

# 🔄 __How Training in a Neural Network Works__

Training is the process of adjusting the **weights** and **biases** so that the neural network makes accurate predictions.

---

## __Step 1: Input Features__

Input features are passed to the **Input Layer**.

---

## __Step 2: Weight Initialization__

Assign **random values** to all weights and biases.

This is called **Weight Initialization**.

Initially, the network does not know anything about the data.

---

## __Step 3: Summation Function__

Each neuron computes

$$
Z = W_1X_1 + W_2X_2 + \cdots + W_nX_n + B
$$

---

## __Step 4: Activation Function__

The weighted sum is passed through an activation function.

$$
A = f(Z)
$$

This introduces **non-linearity** into the network.

---

## __Step 5: Pass Output to Next Layer__

The output of one layer becomes the input to the next layer.

This process continues until the output layer is reached.

---

## __Step 6: Final Prediction__

The output layer produces the final prediction.

---

# ➡️ __Forward Propagation__

Steps **1 to 6** together are called

> **Forward Propagation**

During forward propagation, data moves from the **Input Layer** to the **Output Layer** to generate predictions.

---

## __Step 7: Compute Loss__

Compare the **Predicted Output** with the **Actual Output**.

The difference between them is called the **Loss**.

A smaller loss indicates better predictions.

---

## __Step 8: Update Weights and Biases__

Using **Backpropagation**, calculate the gradients of the loss with respect to each weight and bias.

Then update them to reduce the loss.

---

# ⬅️ __Backward Propagation__

Step **8** is called

> **Backward Propagation**

During backpropagation, gradients are calculated using the **Chain Rule**, and weights are updated using **Gradient Descent**.

---

## __Step 9: Repeat__

Repeat the following steps until the loss becomes sufficiently small:

- Forward Propagation
- Loss Calculation
- Backpropagation
- Weight Update

This iterative process allows the neural network to learn from the training data.

---

# ⚙️ __How Weights and Biases are Updated__

The update equations are

$$
W_{\text{new}}
=
W_{\text{old}}
-
\eta
\frac{\partial J}{\partial W}
$$

$$
b_{\text{new}}
=
b_{\text{old}}
-
\eta
\frac{\partial J}{\partial b}
$$

Where

- **W** → Weight
- **b** → Bias
- **J** → Loss Function
- **η (Eta)** → Learning Rate
- **∂J/∂W** → Gradient of Loss with respect to Weight
- **∂J/∂b** → Gradient of Loss with respect to Bias

---

# 📉 __Gradient Descent__

The optimization algorithm used to update weights is called **Gradient Descent**.

It consists of two terms:

## __Gradient__

The **Gradient** is the derivative of the loss function with respect to the weights.

$$
\frac{\partial J}{\partial W}
$$

It tells us

- Which direction to move.
- How much the loss changes when the weight changes.

---

## __Descent__

Descent means moving **downhill** toward the minimum value of the loss function.

The objective is to find the weights that produce the **minimum loss**.

---

## __Relationship__

```text
Slope
   │
   ▼
Derivative
   │
   ▼
Gradient
   │
   ▼
Gradient Descent
```

---

# 🔄 __Complete Neural Network Training Pipeline__

```text
Input Features
       │
       ▼
Weight Initialization
       │
       ▼
Summation Function
       │
       ▼
Activation Function
       │
       ▼
Forward Propagation
       │
       ▼
Prediction
       │
       ▼
Loss Calculation
       │
       ▼
Backward Propagation
       │
       ▼
Gradient Calculation
       │
       ▼
Update Weights & Biases
       │
       ▼
Repeat Until Loss is Minimum
```

# 📌 __Key Takeaways__

- A neuron performs two operations: **Summation** and **Activation**.
- The summation function computes the weighted sum of inputs and bias.
- Activation functions introduce **non-linearity**, enabling the network to learn complex patterns.
- ReLU is the most commonly used activation function in hidden layers.
- Sigmoid is used for **Binary Classification**.
- Softmax is used for **Multi-class Classification**.
- Linear activation is used for **Regression**.
- A layer is a collection of neurons, and multiple layers form a neural network.
- Forward Propagation computes predictions.
- Backpropagation computes gradients using the **Chain Rule**.
- Gradient Descent updates weights and biases to minimize the loss.
- Training continues until the loss reaches a minimum or an acceptable value.
