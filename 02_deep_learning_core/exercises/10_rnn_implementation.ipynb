## 📚 __What is Sentiment Analysis?__

**Sentiment Analysis** is a Natural Language Processing (**NLP**) task that determines the emotional tone or opinion expressed in a piece of text. The objective is to classify the text into predefined sentiment categories, most commonly:

- **Positive** 😊
- **Negative** 😞

For example:

| **Movie Review** | **Sentiment** |
|:-----------------|:-------------:|
| This movie is amazing. | Positive 😊 |
| The acting was terrible. | Negative 😞 |
| I really enjoyed the film. | Positive 😊 |
| It was boring and slow. | Negative 😞 |

In this notebook, our RNN model will learn patterns from thousands of movie reviews and predict whether a review expresses a **Positive (1)** or **Negative (0)** sentiment.

## 📂 __1. Load the IMDB Dataset__

The **IMDB Movie Review Dataset** is one of the most widely used benchmark datasets for **binary sentiment classification**. It contains movie reviews collected from the Internet Movie Database (IMDB), where each review is labeled according to its sentiment.

### __Dataset Overview__

- **25,000** movie reviews for training.
- **25,000** movie reviews for testing.
- Each review is assigned one of two labels:
  - **1** → Positive Review
  - **0** → Negative Review

### __How is the Dataset Stored?__

Unlike a regular text dataset, the IMDB dataset provided by **Keras** does **not** store reviews as plain English sentences. Instead, each review is already **tokenized** and represented as a sequence of **integer IDs**, where each integer corresponds to a unique word in the dataset's vocabulary.

This numerical representation allows deep learning models, such as **Recurrent Neural Networks (RNNs)**, to process text efficiently.

> **Note:** Later in this notebook, we will learn how these integer IDs represent words and how they can be converted back into readable text.

````python
###############################################
# Import the IMDB Movie Reviews dataset
###############################################
from keras.datasets import imdb

###############################################
# Load the dataset
#
# x_train : Training movie reviews (Input)
# y_train : Training sentiment labels (Output)
#
# x_test  : Testing movie reviews (Input)
# y_test  : Testing sentiment labels (Output)
###############################################
(x_train, y_train), (x_test, y_test) = imdb.load_data()
````

## 📝 __2. NLP Pipeline__

Machine learning and deep learning models cannot understand raw text directly. Before training an RNN, the text must be processed and converted into a numerical format that the model can interpret.

A typical **Natural Language Processing (NLP)** pipeline consists of the following steps:

### __1. Data Collection__

The first step is to gather text data from one or more sources. In this notebook, we will use the **IMDB Movie Review Dataset**, which contains movie reviews labeled as **positive** or **negative**.

### __2. Text Preprocessing__

Raw text often contains unnecessary information that can affect the model's performance. During preprocessing, the text is cleaned and standardized.

Common preprocessing operations include:

- Removing punctuation and special characters.
- Converting text into individual words (**Tokenization**).
- Converting words to lowercase (**Normalization**).
- Removing stop words (optional).
- Stemming or Lemmatization (optional).

### __3. Convert Tokens into Numerical Representation__

Deep learning models work only with numbers. Therefore, each word (or token) is mapped to a unique integer called an **Input ID**. The entire sentence is then represented as a sequence of these integer IDs.

### __4. Train the Model__

Finally, the numerical sequences are provided as input to the RNN. During training, the model learns patterns from the reviews and predicts whether the sentiment is **Positive** or **Negative**.

## 🔤 __Understanding Input IDs__

Deep learning models cannot understand text directly. They can only process **numerical data**. Therefore, every word in a text must first be converted into a unique numerical value before it can be given to the model.

### __What is a Vocabulary?__

A **Vocabulary** is a collection of all the unique words present in a dataset. During preprocessing, each word in the vocabulary is assigned a unique integer value.

These unique integer values are called **Input IDs** (also known as **Token IDs**).

### __Example Vocabulary__

```python
vocab = {
    "movie": 1,
    "is": 2,
    "good": 3,
    "bad": 4,
    "not": 5
}
```

In the above vocabulary:

- `"movie"` → **1**
- `"is"` → **2**
- `"good"` → **3**
- `"bad"` → **4**
- `"not"` → **5**

Each integer uniquely identifies a word in the vocabulary. Instead of processing words, the neural network processes these **Input IDs**, making it possible to perform mathematical computations during training.

## 🔢 __Converting Words into Input IDs__

Once a vocabulary has been created, every word in a sentence is replaced with its corresponding **Input ID**. As a result, each sentence is converted into a sequence of integers.

### __Example__

Suppose we have the following vocabulary:

```python
vocab = {
    "movie": 1,
    "is": 2,
    "good": 3,
    "bad": 4,
    "not": 5
}
```

The sentences are converted as follows:

```
[movie, is, good]
        ↓
[1, 2, 3]

[movie, is, bad]
        ↓
[1, 2, 4]

[movie, is, not, good]
        ↓
[1, 2, 5, 3]
```

Notice that the **order of the words is preserved**. This is important because **Recurrent Neural Networks (RNNs)** process text sequentially, one token at a time. Changing the order of words can completely change the meaning of a sentence.

For example:

```
Movie is good
```

is different from

```
Movie is not good
```

Although the two sentences contain many of the same words, the additional word **"not"** changes the overall sentiment. By preserving the sequence of Input IDs, the RNN can learn such contextual relationships.

## 📖 __Decode Input IDs Back to Words__

So far, we have seen that the IMDB dataset stores each movie review as a sequence of **Input IDs**. While these numerical IDs are suitable for deep learning models, they are not easy for humans to interpret.

To understand the actual content of a review, we need to convert these **Input IDs** back into their corresponding words.

The IMDB dataset provides a **word-to-ID mapping**, where each word in the vocabulary is associated with a unique integer. By reversing this mapping, we can obtain an **ID-to-word mapping**, allowing us to decode integer sequences into readable text.

In the following steps, we will retrieve the vocabulary and use it to identify the words corresponding to different Input IDs.

````python
###############################################
# Get the IMDB vocabulary
#
# Returns a dictionary where:
# Key   -> Word
# Value -> Unique Input ID assigned to the word
###############################################
vocab = imdb.get_word_index()
````

## __Creating a Reverse Vocabulary__

The vocabulary returned by `imdb.get_word_index()` maps **Word → Input ID**.

However, to decode movie reviews, we need the opposite mapping (**Input ID → Word**). Therefore, we create a **reverse vocabulary** by swapping the keys and values of the original dictionary.

This allows us to directly retrieve a word using its corresponding Input ID.

````python
###############################################
# Create a reverse vocabulary
#
# Original Vocabulary:
# Word      -> Input ID
#
# Reverse Vocabulary:
# Input ID  -> Word
###############################################
id_to_word = {index: word for word, index in vocab.items()}

###############################################
# Display the word corresponding to an Input ID
###############################################
input_id = 14

id_to_word[input_id]
````

> **📝 Note**
>
> The IMDB dataset reserves the first few integer IDs for special tokens instead of actual words:
>
> - **0** → Padding (`<PAD>`)
> - **1** → Start of a review (`<START>`)
> - **2** → Unknown word (`<UNK>`)
> - **3** → Unused (`<UNUSED>`)
>
> Because of these reserved tokens, the Input IDs stored in `x_train` are **offset by +3** compared to the IDs returned by `imdb.get_word_index()`.
>
> Therefore, when decoding an actual review from the dataset, this offset must be taken into account.

During model training, reviews are processed in **mini-batches**. For efficient batch processing, all sequences within a batch must have the **same length**.

To achieve this, shorter sequences are padded and longer sequences may be truncated. This process is known as **Padding**, which we will discuss in the next section.

## 📦 __Padding__

**Padding** is the process of making all input sequences the **same length** by adding extra values (called **padding values**) to shorter sequences.

Since movie reviews contain different numbers of words, their corresponding Input ID sequences also have different lengths. Before feeding these sequences into an RNN, they must be converted into a uniform length.

---

### __Why is Padding Required?__

A **Simple RNN** can process sequences of different lengths **one at a time**. However, during training, deep learning models process multiple samples together in **mini-batches** to improve computational efficiency.

For a mini-batch to be processed, every sequence in that batch must have the **same shape (length)**.

Padding ensures that all sequences have a fixed length, allowing them to be grouped into mini-batches and processed efficiently.

---

### __Example__

Suppose the maximum sequence length is **6**.

Before Padding:

```text
[1, 2, 3]
```

After Padding:

```text
[1, 2, 3, 0, 0, 0]
```

Here, **0** is the default **padding value**, which is added until the sequence reaches the desired length.

````python
###############################################
# Import the pad_sequences() function
#
# This function is used to make all input
# sequences the same length by applying
# padding or truncation.
###############################################
from keras.preprocessing.sequence import pad_sequences

###############################################
# Apply padding to the training dataset
#
# maxlen=300
#   -> Set the length of every sequence to 300.
#
# padding="post"
#   -> Add padding values (0) at the end of
#      shorter sequences.
#
# Sequences longer than 300 will be truncated
# according to the default truncating strategy.
###############################################
x_train_padded = pad_sequences(
    x_train,
    maxlen=300,
    padding="post"
)
````

````python
###############################################
# Display the first padded movie review
#
# Observe that the original sequence has been
# extended to the specified length by adding
# padding values (0) at the end.
###############################################
x_train_padded[0]
````

````python
###############################################
# Display the length of the first padded review
#
# Since maxlen=300 was specified, the sequence
# length should now be exactly 300.
###############################################
len(x_train_padded[0])
````

````python
###############################################
# Display the length of another padded review
#
# Notice that every padded review now has the
# same sequence length, regardless of its
# original size.
###############################################
len(x_train_padded[10])
````

## ⚙️ __Padding Hyperparameters__

The `pad_sequences()` function provides several parameters to control how sequences are padded or truncated.

---

### __1. `maxlen`__

Specifies the **desired length** of each sequence after padding.

```python
maxlen = 300
```

- If a sequence is **shorter** than `maxlen`, padding values are added.
- If a sequence is **longer** than `maxlen`, it is truncated.

---

### __2. `padding`__

Determines **where the padding values are added**.

```python
padding = "pre"
```

or

```python
padding = "post"
```

| Option | Description |
|--------|-------------|
| **"pre"** | Add padding at the **beginning** of the sequence *(default)* |
| **"post"** | Add padding at the **end** of the sequence |

**Example**

```text
Original Sequence:
[1, 2, 3]

padding="pre"
[0, 0, 0, 1, 2, 3]

padding="post"
[1, 2, 3, 0, 0, 0]
```

---

### __3. `truncating`__

Determines **which part of a sequence is removed** when it exceeds `maxlen`.

```python
truncating = "pre"
```

or

```python
truncating = "post"
```

| Option | Description |
|--------|-------------|
| **"pre"** | Remove values from the **beginning** of the sequence *(default)* |
| **"post"** | Remove values from the **end** of the sequence |

**Example**

```text
Original Sequence:
[1, 2, 3, 4, 5, 6]

maxlen = 4

truncating="pre"
[3, 4, 5, 6]

truncating="post"
[1, 2, 3, 4]
```

---

### __4. `value`__

Specifies the **padding value** that will be added to shorter sequences.

```python
value = 0
```

- The default padding value is **0**.
- You can specify any other integer if required.

**Example**

```python
pad_sequences(x_train, maxlen=6, value=99)
```

Output:

```text
[1, 2, 3, 99, 99, 99]
```

## 🏗️ __Building the RNN Model__

Now that the input sequences have been converted to a fixed length, we can build and train a **Recurrent Neural Network (RNN)** for sentiment classification.

The overall process of building an RNN consists of the following steps:

### __1. Define the Model__

Create a neural network model that will hold all the layers of the RNN.

### __2. Add the RNN Layer(s)__

Add one or more **SimpleRNN** layers that process the input sequence and learn temporal (sequential) patterns from the text.

### __3. Add the Output Layer__

Add a **Dense** layer that produces the final prediction.

Since sentiment analysis is a **binary classification** problem (Positive or Negative), the output layer contains **one neuron** with the **Sigmoid** activation function.

### __4. Compile the Model__

Configure the model by specifying:

- **Loss Function** – Measures the prediction error.
- **Optimizer** – Updates the model's weights during training.

### __5. Train the Model__

Train the RNN using the training dataset for a specified number of **epochs** so that it can learn patterns and make accurate predictions.

````python
x_train_padded.shape
````

````python
###############################################
# Import the required classes for building
# the Recurrent Neural Network (RNN)
###############################################
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input, SimpleRNN

###############################################
# Create a Sequential model
#
# A Sequential model allows us to stack
# layers one after another.
###############################################
model = Sequential()
````

````python
###############################################
# Add an Input layer
#
# shape=(300, 1)
#   -> 300 time steps (Input IDs)
#   -> 1 feature at each time step
#
# The Input layer defines the shape of a
# single training example that will be
# provided to the RNN.
###############################################
model.add(Input(shape=(300, 1)))

###############################################
# Add a Simple RNN layer
#
# units=10
#   -> Number of neurons (hidden units)
#      in the RNN layer.
#
# The RNN processes the input sequence one
# time step at a time and learns sequential
# patterns from the data.
###############################################
model.add(
    SimpleRNN(units=10)
)
````

````python
###############################################
# Add the output layer
#
# Dense(1)
#   -> One neuron for binary classification.
#
# activation="sigmoid"
#   -> Produces an output between 0 and 1,
#      representing the probability of the
#      review being Positive.
###############################################
model.add(Dense(1, activation="sigmoid"))
````

````python
###############################################
# Compile the model
#
# optimizer="adam"
#   -> Updates the model's weights during
#      training.
#
# loss="binary_crossentropy"
#   -> Loss function used for binary
#      classification problems.
###############################################
model.compile(
    optimizer="adam",
    loss="binary_crossentropy"
)
````

````python
###############################################
# Train the RNN model
#
# x_train_padded : Input sequences
# y_train        : Target labels
# epochs=5       : Train the model for
#                  5 complete passes over
#                  the training dataset.
###############################################
model.fit(
    x_train_padded,
    y_train,
    epochs=5
)
````

## ⚠️ __Why Does a Simple RNN Struggle with Long Sequences?__

One of the major limitations of a **Simple Recurrent Neural Network (RNN)** is its difficulty in learning **long-term dependencies**.

As a Simple RNN processes a sequence, information is passed from one time step to the next through its **hidden state**. Ideally, the hidden state should preserve important information from earlier words so that it can be used later in the sequence.

However, as the sequence becomes longer, information from earlier words gradually becomes weaker. This happens because of the **Vanishing Gradient Problem**, where the network finds it increasingly difficult to preserve information over many time steps.

As a result, the model tends to remember **recent words** better than words that appeared much earlier in the sequence.

---

### __Conceptual Illustration__

Consider the sentence:

```text
"This is a good movie"
```

As the Simple RNN reads the sentence one word at a time, the **current input** is combined with the **previous hidden state** to generate the next hidden state.

```text
h₀ + this     ─────────► h₁      Memory of "this": ██████████ 100%

is + h₁       ─────────► h₂      Memory of "this": ██████░░░░ 60%

good + h₂     ─────────► h₃      Memory of "this": ███░░░░░░░ 30%

movie + h₃    ─────────► h₄      Memory of "this": █░░░░░░░░░ 10%

h₄ ───────────► Output Layer ─────────► Sentiment
```

Here, **h₀** represents the **initial hidden state**, which is typically initialized to all zeros before the first word is processed.

Notice that each new hidden state is computed using:

- **The current input word**
- **The previous hidden state**, which carries information from all previously processed words

As the sequence grows, the hidden state must continuously carry information from earlier words while also incorporating new words. Over many time steps, the influence of the earliest words (such as **"this"**) gradually becomes weaker, while the influence of more recent words becomes stronger.

This is why a **Simple RNN** often struggles to capture **long-term dependencies** in long sequences.

> **📝 Note:** The memory bars and percentages shown above are **only a conceptual illustration** to explain how information from earlier words can gradually weaken in a Simple RNN. They do **not** represent the actual numerical values, hidden state values, memory retention, or weights learned by the network.

## 🚧 __Drawbacks of Simple RNN__

Simple RNNs have several limitations:

- Difficulty learning **long-term dependencies**.
- Suffers from the **Vanishing Gradient Problem**, where gradients become very small during backpropagation.
- May also experience the **Exploding Gradient Problem**, where gradients become excessively large.
- Information from earlier time steps gradually fades, making it difficult to retain long-term context.

These limitations led to the development of more advanced recurrent architectures, such as:

- **LSTM (Long Short-Term Memory)**
- **GRU (Gated Recurrent Unit)**

These architectures use gating mechanisms to retain important information over longer sequences, making them more effective for many sequence learning tasks.
