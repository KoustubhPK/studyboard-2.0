# 🔢 __Text Representation__

- __Vectors:__ array(list) of numbers
- __Embedding:__ Vector representation of text
- __Dimension of Embeddings:__ No of items in list

---

### __Types of Embedding:__
1. __Frequency Based:__
    - One Hot Encoding
    - Bag-of-Words / CountVectorizer
    - TF-IDF

2. __Prediction Based:__
    - Word2Vec
    - Glove
    - Fasttext

````python
review = [
    "movie is good good good", # positive sentiment = 1
    "movie is bad", # negative sentiment = 0
    "movie is not good" # negative sentiment = 0
]

# positive sentiment = 1, negative sentiment = 0
sentiment = [1, 0, 0]
````

````python
import pandas as pd

df = pd.DataFrame(columns=['Review', 'Sentiment'], data=list(zip(review, sentiment)))
df
````

## __1. One Hot Encoding (OHE):__

### __Example Sentences__

```text
Sentence 1: movie is good
            → [movie, is, good]

Sentence 2: movie is bad
            → [movie, is, bad]

Sentence 3: movie is not good
            → [movie, is, not, good]
```

---

### __Vocabulary__

| Index | Word |
|------:|------|
| 0 | movie |
| 1 | is |
| 2 | good |
| 3 | bad |
| 4 | not |

Vocabulary:

```text
[movie, is, good, bad, not]
```

---

### __One-Hot Encoding Matrix__

| Word  | movie | is | good | bad | not |
|--------|:-----:|:--:|:----:|:---:|:---:|
| movie  | 1 | 0 | 0 | 0 | 0 |
| is     | 0 | 1 | 0 | 0 | 0 |
| good   | 0 | 0 | 1 | 0 | 0 |
| bad    | 0 | 0 | 0 | 1 | 0 |
| not    | 0 | 0 | 0 | 0 | 1 |

---

### __One-Hot Representation__

| Word | One-Hot Vector |
|------|----------------|
| movie | [1, 0, 0, 0, 0] |
| is | [0, 1, 0, 0, 0] |
| good | [0, 0, 1, 0, 0] |
| bad | [0, 0, 0, 1, 0] |
| not | [0, 0, 0, 0, 1] |

---

### __Example Representation of Sentences__

```text
movie is good
↓
[
  [1, 0, 0, 0, 0],   # movie
  [0, 1, 0, 0, 0],   # is
  [0, 0, 1, 0, 0]    # good
]

movie is bad
↓
[
  [1, 0, 0, 0, 0],   # movie
  [0, 1, 0, 0, 0],   # is
  [0, 0, 0, 1, 0]    # bad
]

movie is not good
↓
[
  [1, 0, 0, 0, 0],   # movie
  [0, 1, 0, 0, 0],   # is
  [0, 0, 0, 0, 1],   # not
  [0, 0, 1, 0, 0]    # good
]
```

---

### __Pros__

- ✅ **Simple & Intuitive**
  - Easy to understand and implement.

- ✅ **No Training Required**
  - Vectors are created directly from the vocabulary without learning any parameters.

- ✅ **Deterministic Representation**
  - The same word always gets the same unique vector.

---

### __Cons__

- ❌ **No Semantic Meaning**
  - Similar words have completely different vectors.
  - Example: `good` and `excellent` are equally distant as `good` and `car`.

- ❌ **High Dimensionality**
  - Vector length equals the vocabulary size (`|V|`).
  - Larger vocabulary ⇒ larger vectors ⇒ more memory usage.

- ❌ **Sparse Representation**
  - Each vector contains only one `1` and the remaining values are `0`.
  - Sparse vectors waste memory and are computationally inefficient.

- ❌ **Out-of-Vocabulary (OOV) Problem**
  - Words not present in the training vocabulary cannot be represented.
  - Example:
    - Vocabulary = `[movie, is, good, bad, not]`
    - Test sentence = `"movie is excellent"`
    - `excellent` has no one-hot vector.

- ❌ **No Context Information**
  - A word always has the same representation regardless of its context.
  - Example:
    - `bank` in **river bank**
    - `bank` in **bank account**
    - Both receive the same one-hot vector.

- ❌ **No Relationship Between Words**
  - One-hot vectors cannot capture similarity or relationships between words.
  - Example:
    - `king`, `queen`, `man`, and `woman` are treated as completely unrelated.

````python
from sklearn.feature_extraction.text import CountVectorizer

# One-Hot Encoding (Binary Representation)
ohe = CountVectorizer(binary=True)
ohe_vectors = ohe.fit_transform(df['Review']).toarray()

# Vocabulary of the reviews
vocabulary = ohe.get_feature_names_out()
print("Vocabulary:", vocabulary)

ohe_df = pd.DataFrame(ohe_vectors, columns=vocabulary, index=review)
ohe_df
````

# __2. Bag of Words (BoW)__

### __Example Sentences__

```text
Sentence 1: movie is good
Sentence 2: movie is bad
Sentence 3: movie is not good
```

---

### __Vocabulary__

| Index | Word |
|------:|------|
| 0 | movie |
| 1 | is |
| 2 | good |
| 3 | bad |
| 4 | not |

Vocabulary:

```text
[movie, is, good, bad, not]
```
---

### __Bag of Words Representation__

Each sentence is represented by the **frequency (count)** of each word in the vocabulary.

| Sentence | movie | is | good | bad | not |
|----------|:-----:|:--:|:----:|:---:|:---:|
| movie is good | 1 | 1 | 1 | 0 | 0 |
| movie is bad | 1 | 1 | 0 | 1 | 0 |
| movie is not good | 1 | 1 | 1 | 0 | 1 |

---

### __Vector Representation__

| Sentence | BoW Vector |
|----------|------------|
| movie is good | [1, 1, 1, 0, 0] |
| movie is bad | [1, 1, 0, 1, 0] |
| movie is not good | [1, 1, 1, 0, 1] |

---

### __Example__

```text
movie is good
↓
[1, 1, 1, 0, 0]

movie is bad
↓
[1, 1, 0, 1, 0]

movie is not good
↓
[1, 1, 1, 0, 1]
```

> **Note:** BoW only stores the number of occurrences of each word. It ignores the order in which words appear.

---

### __Pros__

- ✅ **Simple & Easy to Implement**
  - Straightforward text representation using word frequencies.

- ✅ **Captures Word Frequency**
  - Important words occurring multiple times receive higher values.

- ✅ **No Training Required**
  - Vectors are generated directly from the corpus.

- ✅ **Works Well for Basic Text Classification**
  - Effective for tasks like spam detection and sentiment analysis on small datasets.

---

### __Cons__

- ❌ **Ignores Word Order**
  - Sentence structure is lost.
  - Example:
    - `dog bites man`
    - `man bites dog`
    - Both produce the same BoW vector.

- ❌ **No Semantic Meaning**
  - Similar words have completely different representations.
  - Example:
    - `good` and `excellent` are treated as unrelated words.

- ❌ **High Dimensionality**
  - Vector size equals the vocabulary size (`|V|`).
  - Large vocabularies produce large vectors.

- ❌ **Sparse Representation**
  - Most values in the vector are `0`, making storage inefficient.

- ❌ **Out-of-Vocabulary (OOV) Problem**
  - New words not present in the vocabulary cannot be represented.
  - Example:
    - Vocabulary = `[movie, is, good, bad, not]`
    - Sentence = `movie is excellent`
    - `excellent` is ignored or treated as unknown.

- ❌ **No Context Information**
  - Words with multiple meanings receive the same representation.
  - Example:
    - `bank` in **river bank**
    - `bank` in **bank account**
    - Both contribute to the same feature.

- ❌ **Cannot Capture Relationships Between Words**
  - It doesn't understand that `king` and `queen` or `good` and `excellent` are semantically related.

- ❌ **Sensitive to Vocabulary Size**
  - As the corpus grows, the vocabulary and feature space become larger, increasing memory and computation requirements.

````python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
bow_vectors = cv.fit_transform(df['Review']).toarray()

# Vocabulary of the reviews
vocabulary = cv.get_feature_names_out()
print("Vocabulary:", vocabulary) # sorted alphabetically

# Bag of Words representation of the reviews
bow_df = pd.DataFrame(bow_vectors, columns=vocabulary, index=review)
bow_df
````

# __3. N-Gram Bag of Words__

An **N-Gram** is a sequence of **N consecutive words**.

- **Unigram (N = 1)** → Single words
- **Bigram (N = 2)** → Two consecutive words
- **Trigram (N = 3)** → Three consecutive words

---

### __Example Sentences__

```text
Sentence 1: movie is good
Sentence 2: movie is bad
Sentence 3: movie is not good
```

---

## __3.1 Unigram (N = 1)__

### __Vocabulary__

```text
[movie, is, good, bad, not]
```

---

### __Unigram BoW Matrix__

| Sentence | movie | is | good | bad | not |
|----------|:-----:|:--:|:----:|:---:|:---:|
| movie is good | 1 | 1 | 1 | 0 | 0 |
| movie is bad | 1 | 1 | 0 | 1 | 0 |
| movie is not good | 1 | 1 | 1 | 0 | 1 |

---

## __3.2 Bigram (N = 2)__

### __Generate Bigrams__

```text
Sentence 1:
movie is
is good

Sentence 2:
movie is
is bad

Sentence 3:
movie is
is not
not good
```
---

### __Vocabulary__

```text
[
movie is,
is good,
is bad,
is not,
not good
]
```
---

### __Bigram BoW Matrix__

| Sentence | movie is | is good | is bad | is not | not good |
|----------|:---------:|:-------:|:------:|:------:|:---------:|
| movie is good | 1 | 1 | 0 | 0 | 0 |
| movie is bad | 1 | 0 | 1 | 0 | 0 |
| movie is not good | 1 | 0 | 0 | 1 | 1 |

---

## __3.3 Trigram (N = 3)__

### __Generate Trigrams__

```text
Sentence 1:
movie is good

Sentence 2:
movie is bad

Sentence 3:
movie is not
is not good
```

---

### __Vocabulary__

```text
[
movie is good,
movie is bad,
movie is not,
is not good
]
```

---

### __Trigram BoW Matrix__

| Sentence | movie is good | movie is bad | movie is not | is not good |
|----------|:-------------:|:------------:|:------------:|:------------:|
| movie is good | 1 | 0 | 0 | 0 |
| movie is bad | 0 | 1 | 0 | 0 |
| movie is not good | 0 | 0 | 1 | 1 |

---

### __Pros__

- ✅ **Captures Word Order**
  - Unlike standard BoW, N-Grams preserve the order of consecutive words.

- ✅ **Captures Local Context**
  - Phrases such as `not good`, `very good`, and `machine learning` are treated as meaningful features.

- ✅ **Improves Text Classification**
  - Often performs better than Unigram BoW for sentiment analysis and document classification.

- ✅ **Simple to Implement**
  - Easy extension of the Bag of Words model.

---

### __Cons__

- ❌ **Vocabulary Size Grows Rapidly**
  - Higher values of **N** generate many more unique features.
  - Example:
    - Unigram → 10,000 features
    - Bigram → 100,000+ features
    - Trigram → Millions of possible features

- ❌ **High Dimensionality**
  - Larger vocabularies require more memory and computation.

- ❌ **Sparse Representation**
  - Most N-Grams occur rarely, producing vectors with many zeros.

- ❌ **Out-of-Vocabulary (OOV) Problem**
  - Unseen N-Grams cannot be represented.
  - Example:
    - Training Bigram: `is good`
    - Test Bigram: `looks good`
    - `looks good` is unknown.

- ❌ **Limited Context**
  - Only captures relationships within **N consecutive words**.
  - Long-range dependencies are ignored.

- ❌ **Data Sparsity Increases with N**
  - Many higher-order N-Grams appear only once, making learning difficult.

- ❌ **Still No Semantic Understanding**
  - Similar phrases are treated as completely different.
  - Example:
    - `very good`
    - `really good`
    - These are different N-Grams even though they convey similar meaning.

---

### __Choosing N__

| N-Gram | Captures | Advantages | Disadvantages |
|--------|----------|------------|---------------|
| **Unigram (N=1)** | Individual words | Small vocabulary, fast | Ignores word order |
| **Bigram (N=2)** | Two-word phrases | Captures local context | Larger vocabulary |
| **Trigram (N=3)** | Three-word phrases | Better phrase representation | Very high dimensional and sparse |

````python
# n-gram range: default is (1, 1) which means unigrams only. We can change it to (1, 2) to include bigrams as well.
from sklearn.feature_extraction.text import CountVectorizer
````

````python
# unigram representation of the reviews

cv = CountVectorizer(ngram_range=(1,1)) # unigrams, bigrams, and trigrams
unigram_vectors = cv.fit_transform(df['Review']).toarray()

# Vocabulary of the reviews
vocabulary = cv.get_feature_names_out()
print("Vocabulary:", vocabulary) # sorted alphabetically

bow_df = pd.DataFrame(unigram_vectors, columns=vocabulary, index=review)
bow_df
````

````python
# bigrams representation of the reviews

cv = CountVectorizer(ngram_range=(1,2)) # unigrams, bigrams
bigram_vectors = cv.fit_transform(df['Review']).toarray()

# Vocabulary of the reviews
vocabulary = cv.get_feature_names_out()
print("Vocabulary:", vocabulary) # sorted alphabetically

bow_df = pd.DataFrame(bigram_vectors, columns=vocabulary, index=review)
bow_df
````

````python
# trigrams representation of the reviews

cv = CountVectorizer(ngram_range=(1,3)) # unigrams, bigrams, and trigrams
trigram_vectors = cv.fit_transform(df['Review']).toarray()

# Vocabulary of the reviews
vocabulary = cv.get_feature_names_out()
print("Vocabulary:", vocabulary) # sorted alphabetically

bow_df = pd.DataFrame(trigram_vectors, columns=vocabulary, index=review)
bow_df
````

# __4. TF-IDF (Term Frequency - Inverse Document Frequency)__

TF-IDF assigns a weight to each word based on:
- **Term Frequency (TF):** How often a word appears in a document.
- **Inverse Document Frequency (IDF):** How rare or important the word is across all documents.

Words that occur frequently in a document but rarely across the corpus receive **higher weights**.

---

### __Example Documents__

```text
Document 1: movie is good
Document 2: movie is bad
Document 3: movie is not good
```

Total Documents (**N**) = **3**

---

### __Vocabulary__

```text
[movie, is, good, bad, not]
```

---

# __Step 1: Calculate Term Frequency (TF)__

### __Formula__

```text
TF = (Number of times a word appears in a document)
     ------------------------------------------------
        Total number of words in the document
```

### __TF Matrix__

| Word | D1 | D2 | D3 |
|------|:--:|:--:|:--:|
| movie | 1/3 = 0.33 | 1/3 = 0.33 | 1/4 = 0.25 |
| is | 1/3 = 0.33 | 1/3 = 0.33 | 1/4 = 0.25 |
| good | 1/3 = 0.33 | 0 | 1/4 = 0.25 |
| bad | 0 | 1/3 = 0.33 | 0 |
| not | 0 | 0 | 1/4 = 0.25 |

---

# __Step 2: Calculate Document Frequency (DF)__

Document Frequency = Number of documents containing the word.

| Word | DF |
|------|:--:|
| movie | 3 |
| is | 3 |
| good | 2 |
| bad | 1 |
| not | 1 |

---

# __Step 3: Calculate Inverse Document Frequency (IDF)__

### __Formula__

```text
IDF = log(N / DF)
```

where

- N = Total number of documents
- DF = Document Frequency

### __IDF Values__

| Word | DF | IDF = log(3 / DF) |
|------|:--:|:-----------------:|
| movie | 3 | 0 |
| is | 3 | 0 |
| good | 2 | 0.176 |
| bad | 1 | 0.477 |
| not | 1 | 0.477 |


> **Note:** Values are approximate using log base 10. Different libraries may use natural logarithm or smoothing, resulting in slightly different values.

---

# __Step 4: Calculate TF-IDF__

### __Formula__

```text
TF-IDF = TF × IDF
```

---

### __TF-IDF Matrix__

| Word | D1 | D2 | D3 |
|------|:--:|:--:|:--:|
| movie | 0 | 0 | 0 |
| is | 0 | 0 | 0 |
| good | 0.33 × 0.176 = 0.058 | 0 | 0.25 × 0.176 = 0.044 |
| bad | 0 | 0.33 × 0.477 = 0.157 | 0 |
| not | 0 | 0 | 0.25 × 0.477 = 0.119 |

---

### __Final TF-IDF Vectors__

| Document | TF-IDF Vector |
|----------|---------------|
| movie is good | [0, 0, 0.058, 0, 0] |
| movie is bad | [0, 0, 0, 0.157, 0] |
| movie is not good | [0, 0, 0.044, 0, 0.119] |

---

### __Interpretation__

- **movie** and **is** appear in every document.
  - Their IDF is **0**, so they receive **0 weight**.

- **bad** appears in only one document.
  - It receives the **highest TF-IDF score**.

- **good** appears in two documents.
  - It receives a **moderate weight**.

- **not** appears in only one document.
  - It also receives a **high TF-IDF score**.

Thus, TF-IDF automatically reduces the importance of common words while emphasizing informative words.

---

### __Pros__

- ✅ **Captures Word Importance**
  - Assigns higher weights to informative words.

- ✅ **Reduces Effect of Common Words**
  - Frequently occurring words receive lower weights.

- ✅ **Simple and Efficient**
  - Easy to compute and widely used in NLP.

- ✅ **Works Well for Information Retrieval**
  - Used in search engines and document ranking.

- ✅ **Better than Bag of Words**
  - Uses weighted frequencies instead of raw counts.

---

### __Cons__

- ❌ **No Semantic Meaning**
  - Similar words receive different representations.
  - Example:
    - `good` and `excellent` are treated as unrelated.

- ❌ **Ignores Word Order**
  - Sentence structure is lost.
  - Example:
    - `dog bites man`
    - `man bites dog`
    - Can produce similar feature vectors.

- ❌ **Sparse Representation**
  - Most entries in the TF-IDF matrix are zero.

- ❌ **Out-of-Vocabulary (OOV) Problem**
  - Words not seen during training cannot be represented.

- ❌ **High Dimensionality**
  - Vector size grows with vocabulary size.

- ❌ **No Context Information**
  - A word always has the same weight regardless of its meaning in different contexts.

- ❌ **Requires Entire Corpus**
  - IDF cannot be computed accurately without access to all documents.

---

### __Applications__

- Search Engines
- Information Retrieval
- Document Ranking
- Keyword Extraction
- Document Similarity
- Text Classification
- Recommendation Systems

````python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_vectors = tfidf.fit_transform(df['Review']).toarray()

# vocabulary of the reviews
vocabulary = tfidf.get_feature_names_out()
print("Vocabulary:", vocabulary) # sorted alphabetically

tfidf_df = pd.DataFrame(tfidf_vectors, columns=vocabulary, index=review)
tfidf_df
````
