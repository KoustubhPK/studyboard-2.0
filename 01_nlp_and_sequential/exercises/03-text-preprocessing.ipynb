# 🧹 __Text Preprocessing__

⚙️ **NLP Pipeline:** It is an iterative process, not a linear one.

- Data Gathering (Text Data)
- Text Preprocessing
    - Text Cleaning
    - Tokenization
    - Text Normalization
- Text Representation (Text → Numbers)
- Model Building
- Model Evaluation
- Model Deployment

## 🔤 __1. Text Normalization__

- Lowercasing:
    - text, Text, TEXT → text (making text uniform)
- Avoid lowercasing when:
    - Named Entity Recognition (NER)
    - Sometimes in Sentiment Analysis (capitalization may carry sentiment)

## 🪓 __2. Stemming__

- Reduces words to their root/base form called a **stem**.
- **Overstemming:** Two or more unrelated words are reduced to the same stem.
    - Example: *universe* and *university* may be reduced to a similar stem by aggressive stemmers.
- **Understemming:** Two or more related words are reduced to different stems.
    - Example: *walking* → *walk*, *walks* → *walks* (not reduced further by some stemmers).
- **Common Stemming Algorithms:**
    1. **Porter Stemmer** – Supports only the English language.
    2. **Snowball Stemmer** – Supports multiple languages and is an improved version of Porter Stemmer.
    3. **Lancaster Stemmer** – An aggressive stemmer.

```python
from nltk.stem import PorterStemmer, SnowballStemmer, LancasterStemmer
```

```python
print(SnowballStemmer.languages)
```

```python
porter = PorterStemmer()
lancaster = LancasterStemmer()
snowball = SnowballStemmer("english")
```

```python
word = "news"

print(f"{'Porter Stemmer':<20}: {porter.stem(word)}")
print(f"{'Snowball Stemmer':<20}: {snowball.stem(word)}")
print(f"{'Lancaster Stemmer':<20}: {lancaster.stem(word)}")
```

## 🌱 __3. Lemmatization__

- Converts words to their dictionary/base form called a **lemma**.
- **Difference between Stemming and Lemmatization:**
    - Stemming simply removes prefixes/suffixes using heuristic rules.
    - Lemmatization converts a word to its dictionary form using vocabulary and morphological analysis.

### __Examples__

| Word | Stemming | Lemmatization |
|------|----------|---------------|
| studies | studi | study |
| running | run | run |
| better | better | good |
| mice | mice | mouse |
