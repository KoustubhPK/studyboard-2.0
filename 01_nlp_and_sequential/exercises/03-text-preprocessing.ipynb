## 🪓 __2. Stemming__

- Reduces words to their root/base form called a **stem**.
- **Overstemming:** Two or more unrelated words are reduced to the same stem.
    - Example: *universe* and *university* may be reduced to a similar stem by aggressive stemmers.
- **Understemming:** Two or more related words are reduced to different stems.
    - Example: *walking* → *walk*, *walks* → *walks* (not reduced further by some stemmers).
- **Common Stemming Algorithms:**
    1. **Porter Stemmer** – Supports only the English language.
    2. **Snowball Stemmer** – Supports multiple languages and is an improved version of Porter Stemmer.
    3. **Lancaster Stemmer** – An aggressive stemmer.

```python
from nltk.stem import PorterStemmer, SnowballStemmer, LancasterStemmer
```

```python
porter = PorterStemmer()
lancaster = LancasterStemmer()
snowball = SnowballStemmer("english")
```

## 🌱 __3. Lemmatization__

- Converts words to their dictionary/base form called a **lemma**.
- Uses a vocabulary (lexicon) and morphological analysis to determine the correct base form.
- In NLTK, lemmatization commonly uses **WordNet** (a lexical database of English) to find the correct lemma.
- **Difference between Stemming and Lemmatization:**
    - Stemming simply removes prefixes/suffixes using heuristic rules.
    - Lemmatization converts a word to its dictionary form using vocabulary and morphological analysis.

### __Examples__

| Word | Stemming | Lemmatization |
|------|----------|---------------|
| studies | studi | study |
| running | run | run |
| better | better | good |
| mice | mice | mouse |

---

### ❓ __Why Stemming?__

- **Simplicity:** Uses simple heuristic rules to remove prefixes/suffixes, making it easy to implement.
- **Efficient:** Faster and requires less computational resources since it does not perform vocabulary lookup or morphological analysis.

---

### 🌿 __When to Prefer Stemming over Lemmatization?__

- **Early Stages:** Useful during the initial stages of text preprocessing when a quick normalization of words is sufficient.
- **Large Datasets:** Preferred when working with large datasets where processing speed is more important than obtaining linguistically correct base forms.
- **Information Retrieval:** Commonly used in search engines and document retrieval systems where exact dictionary words are not essential.
- **When Speed Matters:** Choose stemming if computational efficiency is a higher priority than accuracy.

---

### 🚫 __Where to Avoid Stemming & Lemmatization__

- **Where grammatical structure is important**, because reducing words to their root/base form may change the meaning or grammar.
- **Examples:**
  - Machine Translation
  - Text Generation
  - Grammar Checking
  - Summarization (when preserving sentence structure, grammar and tense are important)

> **Example:**
> - "is", "are", "was" → "be"
> - This can distort tense and grammatical information needed for translation.

---

### ✅ __Where to Use Stemming & Lemmatization__

These techniques are useful when the meaning of the word is more important than its grammatical form.

- **Keyword Extraction**
- **Information Retrieval (Search Engines)**
- **Topic Modeling**
- **Document Classification**
- **Text Classification (Spam Detection, Sentiment Analysis)**
- **Document Clustering**
- **Search and Indexing**

---

> **Rule of Thumb:**
> - **Stemming →** Faster, simpler, less accurate, May produce non-dictionary words.
> - **Lemmatization →** Slower, more accurate, returns valid dictionary words (often using **WordNet**).

```python
# from nltk download wordnet
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
word = "studies"
print(f"{'WordNet Lemmatizer (noun)':<35}: {lemmatizer.lemmatize(word)}")               # default pos is noun, so it will return "study"
print(f"{'WordNet Lemmatizer (adjective)':<35}: {lemmatizer.lemmatize(word,pos='a')}")  # pos = adjective, so it will return "studies" instead of "study"
```

```python
word = "running"
print(f"{'WordNet Lemmatizer (noun)':<35}: {lemmatizer.lemmatize(word)}")          # default pos is noun, so it will return "running" instead of "run"
print(f"{'WordNet Lemmatizer (verb)':<35}: {lemmatizer.lemmatize(word, pos='v')}") # pos = verb, so it will return "run"
```

# 🗑️ __4. Stopwords__

- **Stop words** are commonly occurring words that carry **little or no meaningful information** in many NLP tasks.
- They are often removed during text preprocessing to reduce noise and improve efficiency.

### __Examples__

| Stop Words | Meaningful Words |
|------------|------------------|
| the | machine |
| is | learning |
| a | python |
| and | algorithm |
| of | data |
| in | analysis |

---

> **Example:**
>
> **Original:** *The cat is sitting on the mat.*
>
> **After Stop Word Removal:** *cat sitting mat*

---

### 🛑 __Why Remove Stop Words?__

- **Reduces noise** in the text.
- **Decreases vocabulary size.**
- **Improves processing speed.**
- **Focuses on important words** that carry meaning.
- Can improve the performance of many NLP models.

---

### ✅ __Where to Use Stop Word Removal__

- Information Retrieval (Search Engines)
- Keyword Extraction
- Topic Modeling
- Document Classification
- Text Classification (Spam Detection, Sentiment Analysis)
- Text Clustering
- Search & Indexing

---

### 🚫 __Where NOT to Remove Stop Words__

Removing stop words may change the meaning of a sentence.

- Machine Translation
- Text Summarization
- Question Answering
- Chatbots & Conversational AI
- Grammar Checking
- Sentiment Analysis *(when words like **not**, **no**, **never** are important)*

> **Example:**
>
> **Original:** *This movie is **not** good.*
>
> **After removing "not":** *This movie is good.* ❌ (Meaning changes completely)

---

### 📝 __Custom Stop Words__

- Besides the predefined stop words, we can create our **own custom stop words** based on the dataset.
- Example:
  - In a movie review dataset, words like **movie**, **film**, or **cinema** may appear in almost every document and contribute little to distinguishing reviews.
  - These can be added to the stop word list if appropriate.

---

### 💡 __Rule of Thumb__

- ✅ Remove stop words when they **do not affect the meaning** of the text and your goal is efficient text analysis.
- ❌ Keep stop words when **context, grammar, or sentence meaning** is important.

```python
# from nltk download stopwords
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

print(stopwords.fileids())  # print available stopwords languages
print(stopwords.words('english')[:10])  # print first 10 stopwords
```

## __4.1 Removing unnecessary StopWord from existing words list__

```python
# removing stopwords from a list of words
english_stops = stopwords.words('english')

if "not" in english_stops:  # check if "not" is a stopword
    print("'not' is a stopword")

english_stops.remove("not")  # remove "not" from stopwords list
print("'not' is a stopword" if "not" in english_stops else "'not' is not a stopword anymore")
```

## __4.2 Adding domain specific StopWord in existing words list__

```python
english_stops.append("domain")  # add "domain" word to stopwords list

if "domain" in english_stops:  # check if "domain" is a stopword
    print("'domain' is a stopword")
```

## __4.3 Removing StopWord from Text__

```python
text = "This is a sample text with some stopwords. We will remove the stopwords from this text."

words = text.split()  # split the text into words
filtered_words = [word for word in words if word.lower() not in english_stops]  # remove stopwords

print("Original words:", words)
print("Filtered words:", filtered_words)
```

```python
# creating function to remove stopwords from a text
def remove_stopwords(text, stopwords_list):
    words = text.split()  # split the text into words
    filtered_words = [word for word in words if word.lower() not in stopwords_list]  # remove stopwords
    return " ".join(filtered_words)  # join the filtered words back into a string
```

```python
text = "The sun is shining in the sky. There is not a cloud in sight."

filtered_text = remove_stopwords(text, english_stops)
print("Original text:", text)
print("Filtered text:", filtered_text)
```
