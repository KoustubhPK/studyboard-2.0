# __1. Tokenization__
### __Basic Terms__
- __Tokenization:__ Splitting text into basic unit
- __Token:__ Unit of text
- __corpus :__ All available data
- __Document :__ Each individual sample
- __Vocabulary :__ All unique tokens in a corpus
- __Out-of-Vocabulary (OOV) :__ tokens that are not available in vocab

```python
# Splitting a string into a list of tokens in simple pythonic way.
text = "All unique tokens in a corpus."
text.split()
```

### __NLTK: Natural Language Toolkit__
- Python library for NLP
- Installation: `pip install nltk`

```python
text = "Ice-cream will cost    $2.10 in New Delhi."
```

```python
## whitespace tokenizer : split on white spaces
from nltk.tokenize import WhitespaceTokenizer

tokenizer = WhitespaceTokenizer()
tokens = tokenizer.tokenize(text)
print(tokens)
```

```python
## wordpunct : split on spaces & punctuations
from nltk.tokenize import WordPunctTokenizer

tokenizer = WordPunctTokenizer()
tokens = tokenizer.tokenize(text)
print(tokens)
```

- `word_tokenize()` needs extra dependancy, download the `punkt_tab` package from `nltk`

```python
# from nltk download punkt_tab package for word_tokenize.
import nltk
nltk.download('punkt_tab')
```

```python
## Word tokenizer : split on words
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text)
print(tokens)
```

```python
## Sentence tokenizer : split on sentences
from nltk.tokenize import sent_tokenize

text = "Ice-cream will 'cost' $2.10 in New Delhi? yes, it will"
sentences = sent_tokenize(text)
print(sentences)
```

# __2. Sub-Word Tokenizer__
- Sub-Word tokenization eliminates OOV problem.

    - happiness     --> happi, ness _(seen)_
    - unhappiness   --> un, happi, ness _(seen)_
    - pleasant      --> pleasant _(seen)_
    - `unpleasant    --> un, pleasant` _(unseen)_

- Methods of sub-word tokenization :
    - Byte-pair Encoding (BPE) : GPT, GPT-2, RoBERTa
    - WordPiece : BERT, DistilBERT
    - SentencePiece : BART, T5

---

# __2.1. Byte-Pair Encoding (BPE)__
_Byte-pair Encoding (BPE): GPT, GPT-2, RoBERTa | GPT (40,478)  --> 478 (base vocab) + 40,000 (Merges)_
- __Step 1:__ Pre-tokenization (Split text into words)
- __Step 2:__ Create base vocabulary (with all unique characters)
- __Step 3:__ Merge most frequently occurring pair of characters & add to vocab
- __Step 4:__ Repeat step 3, till desired vocabulary size is reached.

---

|            | Base Vocab       | 1st Merge      | 2nd Merge    | 3rd Merge  |
|------------|------------------|----------------|--------------|------------|
| bat (10)   | b, a, t          | b, at          | b, at        | b, at      |
| cat (5)    | c, a, t          | c, at          | c, at        | c, at      |
| man (10)   | m, a, n          | m, a, n        | ma, n        | man        |
| batman (6) | b, a, t, m, a, n | b, at, m, a, n | b, at, ma, n | b, at, man |
| dog (5)    | d, o, g          | d, o, g        | d, o, g      | d, o, g    |

---

- vocab = [b, a, t, c, m, n, d, o, g] `base Vocab`
- vocab = [b, a, t, c, m, n, d, o, g, at] `1st merge`
- vocab = [b, a, t, c, m, n, d, o, g, at, ma] `2nd merge`
- vocab = [b, a, t, c, m, n, d, o, g, at, ma, man] `3rd merge`

# __2.2. WordPiece__
_WordPiece: BERT, DistilBERT_
- __Step 1:__ Pre-tokenization (Split text into words)
- __Step 2:__ Create base vocabulary (with all unique characters)
- __Step 3:__ Merge:
    - Two character based score, add it to vocab
    - Score = `Frequency of Pair Togather / Frequency of 1st Character * Frequency of 2nd Character`
- __Step 4:__ Repeat step 3, till desired vocabulary size is reached.

---

|            | Base Vocab       | 1st Merge      | 2nd Merge    | 3rd Merge  |
|------------|------------------|----------------|--------------|------------|
| bat (10)   | b, a, t          | b, at          | b, at        | b, at      |
| cat (5)    | c, a, t          | c, at          | c, at        | c, at      |
| man (10)   | m, a, n          | m, a, n        | ma, n        | man        |
| batman (6) | b, a, t, m, a, n | b, at, m, a, n | b, at, ma, n | b, at, man |
| dog (5)    | d, o, g          | d, o, g        | d, o, g      | d, o, g    |

- __Score(a, t) = Frequency Pair(at) / (Frequency(a) × Frequency(t))__
    - __= 21 / (37 × 21)__

# __2.3. SentencePiece__
_SentencePiece: BART, T5, ALBERT, XLNet_
- __Step 1:__ Use raw text directly (no pre-tokenization into words).
- __Step 2:__ Replace each whitespace with the special whitespace symbol ▁ (U+2581), then create the base vocabulary.
- __Step 3:__ Merge the most frequent pair of symbols and add it to the vocabulary.
- __Step 4:__ Repeat step 3, till desired vocabulary size is reached.

---

__Pre-tokenization in BPE / Wordpiece:__
- __Input:__ ``"For Some languages space is also important"``
- __Output:__ ``[For, Some, languages, space, is, also, important]``

__Input processing in SentencePiece (no pre-tokenization):__
- __Input:__ ``"For Some languages space is also important"``
- __Transformed Input:__ ``"▁For▁Some▁languages▁space▁is▁also▁important"``

__Then, after the SentencePiece model tokenizes it, we might get something like:__
- ["▁For", "▁Some", "▁language", "s", "▁space", "▁is", "▁also", "▁important"] or
- ["▁For", "▁Some", "▁languages", "▁space", "▁is", "▁also", "▁important"]

__Note:__
- SentencePiece supports both `BPE` and `Unigram` algorithms. The above steps describe the BPE variant.
- Output depending on the learned vocabulary.
- SentencePiece learns its vocabulary only from the training corpus (the text you provide). It does not use:
    - ❌ WordNet _A lexical database (dictionary of words, synonyms, antonyms, hypernyms, etc.)._
    - ❌ Dictionaries
    - ❌ Linguistic rules
    - ❌ Part-of-speech tags

# 💡 __Summary__

#### __1. Tokenization__
- `.split()` → Split text using whitespace.
- `WhitespaceTokenizer()` → Tokenize based only on whitespace.
- `WordPunctTokenizer()` → Split words and punctuation separately.
- `word_tokenize()` → General-purpose word tokenizer.
- `sent_tokenize()` → Split text into sentences.

---

#### __2. Subword Tokenization__

**1. Byte-Pair Encoding (BPE)**
- Pre-tokenizes text into words.
- Merge the most frequent pair of characters/subwords.
- Used in: GPT, GPT-2, RoBERTa.

**2. WordPiece**
- Pre-tokenizes text into words.
- Merge the pair having the highest score.
- Score = Frequency(pair) / (Frequency(first token) × Frequency(second token))
- Used in: BERT, DistilBERT.

**3. SentencePiece**
- No pre-tokenization (works directly on raw text).
- Represents whitespace using `▁`.
- Learns subword tokens from raw text (using BPE or Unigram).
- Used in: T5, BART, ALBERT, XLNet.

| Feature                  | BPE                 | WordPiece        | SentencePiece           |
|--------------------------|---------------------|------------------|-------------------------|
| Pre-tokenization         | ✅ Yes              | ✅ Yes          | ❌ No                   |
| Merge Criterion          | Most frequent pair  | Highest score    | BPE or Unigram          |
| Uses whitespace symbol ▁ | ❌                   | ❌                | ✅                   |
| Works on raw text        | ❌                   | ❌                | ✅                   |
| Example Models           | GPT, GPT-2, RoBERTa | BERT, DistilBERT | T5, BART, ALBERT, XLNet |
