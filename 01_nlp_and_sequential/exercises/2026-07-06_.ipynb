# __1. Tokenization__
### __Basic Terms__
- __Tokenization:__ Splitting text into basic unit
- __Token:__ Unit of text
- __corpus :__ All available data
- __Document :__ Each individual sample
- __Vocabulary :__ All unique tokens in a corpus
- __Out-of-Vocabulary (OOV) :__ tokens that are not available in vocab

```python
# Splitting a string into a list of tokens in simple pythonic way.
text = "All unique tokens in a corpus."
text.split()
```

### __NLTK: Natural Language Toolkit__
- Python library for NLP
- Installation: `pip install nltk`

```python
text = "Ice-cream will cost    $2.10 in New Delhi."
```

```python
## whitespace tokenizer : split on white spaces
from nltk.tokenize import WhitespaceTokenizer

tokenizer = WhitespaceTokenizer()
tokens = tokenizer.tokenize(text)
print(tokens)
```

```python
## wordpunct : split on spaces & punctuations
from nltk.tokenize import WordPunctTokenizer

tokenizer = WordPunctTokenizer()
tokens = tokenizer.tokenize(text)
print(tokens)
```

- `word_tokenize()` needs extra dependancy, download the `punkt_tab` package from `nltk`

```python
# from nltk download punkt_tab package for word_tokenize.
import nltk
nltk.download('punkt_tab')
```

```python
## Word tokenizer : split on words
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text)
print(tokens)
```

```python
## Sentence tokenizer : split on sentences
from nltk.tokenize import sent_tokenize

text = "Ice-cream will 'cost' $2.10 in New Delhi? yes, it will"
sentences = sent_tokenize(text)
print(sentences)
```

# __2. Sub-Word Tokenizer__
- Sub-Word tokenization eliminates OOV problem.

    - happiness     --> happi, ness _(seen)_
    - unhappiness   --> un, happi, ness _(seen)_
    - pleasant      --> pleasant _(seen)_
    - `unpleasant    --> un, pleasant` _(unseen)_

- Methods of sub-word tokenization :
    - Byte-pair Encoding (BPE) : GPT, GPT-2, RoBERTa
    - WordPiece : BERT, DistilBERT
    - SentencePiece : BART, T5

---

# __2.1. Byte-Pair Encoding _(BPE)___
- __Step 1:__ Pre-tokenization (Split text into words)
- __Step 2:__ Create base vocabulary (with all unique characters)
- __Step 3:__ Merge most frequently occuring pair of characters & add to vocab
- __Step 4:__ Repeat step 3, till desired vocabulary size is reached.

---

|            | Base Vocab       | 1st Merge      | 2nd Merge    | 3rd Merge  |
|------------|------------------|----------------|--------------|------------|
| bat (10)   | b, a, t          | b, at          | b, at        | b, at      |
| cat (5)    | c, a, t          | c, at          | c, at        | c, at      |
| man (10)   | m, a, n          | m, a, n        | ma, n        | man        |
| batman (6) | b, a, t, m, a, n | b, at, m, a, n | b, at, ma, n | b, at, man |
| dog (5)    | d, o, g          | d, o, g        | d, o, g      | d, o, g    |

---

- vocab = [b, a, t, c, m, n, d, o, g] `base Vocab`
- vocab = [b, a, t, c, m, n, d, o, g, at] `1st merge`
- vocab = [b, a, t, c, m, n, d, o, g, at, ma] `2nd merge`
- vocab = [b, a, t, c, m, n, d, o, g, at, ma, man] `3rd merge`
