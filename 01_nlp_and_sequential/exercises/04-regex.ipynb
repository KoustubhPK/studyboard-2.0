# 🔍 __1. Regex__

### __1.1 What is Regex?__

- **Regular Expression (Regex)** is a sequence of characters that defines a **search pattern**.
- It is used to **search, match, extract, replace, and validate** text.
- Regex is supported by many programming languages, including **Python**, **Java**, **JavaScript**, **C++**, and **R**.

---

### __1.2 Why Use Regex?__

Regex helps in:

- 🔍 Finding specific patterns in text.
- ✂️ Extracting useful information from text.
- 🔄 Replacing or modifying text.
- ✅ Validating input such as email addresses, phone numbers, URLs, and passwords.
- 🧹 Cleaning and preprocessing text for NLP.

---

### __1.3 Common Applications__

- Email validation
- Phone number validation
- URL extraction
- Password validation
- Log file analysis
- Data cleaning
- Web scraping
- Text preprocessing in NLP

---

### __1.4 Python `re` Module__

Python provides the built-in **`re`** module to work with regular expressions.

```python
import re
```

---

## 📌 __2. Common Character Classes__

| Pattern | Description | Example Match |
|---------|-------------|---------------|
| `.` | Any character except a newline (`\n`) | `a`, `7`, `#` |
| `\d` | Any digit (`0-9`) | `5`, `9` |
| `\D` | Any character that is **not** a digit | `a`, `@`, `_` |
| `\w` | Word character (`a-z`, `A-Z`, `0-9`, `_`) | `A`, `7`, `_` |
| `\W` | Any character that is **not** a word character | `@`, `#`, ` ` |
| `\s` | Whitespace character (space, tab, newline) | ` `, `\t`, `\n` |
| `\S` | Any character that is **not** whitespace | `A`, `5`, `@` |

---

## 📌 __3. Anchors__

| Pattern | Description | Example |
|---------|-------------|---------|
| `\b` | Word boundary | `\bcat\b` matches **cat**, not **category** |
| `\B` | Not a word boundary | `\Bcat` matches **scat**, not **cat** |
| `^` | Beginning of a string | `^Hello` |
| `$` | End of a string | `world$` |

---

## 📌 __4. Character Sets & Groups__

| Pattern | Description | Example |
|---------|-------------|---------|
| `[abc]` | Matches any one character inside the brackets | `a`, `b`, or `c` |
| `[^abc]` | Matches any character **not** inside the brackets | `d`, `1`, `@` |
| `\|` | OR operator | `cat\|dog` |
| `( )` | Groups expressions together | `(ab)+` |

---

## 📌 __5. Quantifiers__

| Pattern | Description | Example |
|---------|-------------|---------|
| `*` | Matches **0 or more** occurrences | `ab*` → `a`, `ab`, `abb` |
| `+` | Matches **1 or more** occurrences | `ab+` → `ab`, `abb` |
| `?` | Matches **0 or 1** occurrence | `ab?` → `a`, `ab` |
| `{3}` | Matches **exactly 3** occurrences | `a{3}` → `aaa` |
| `{3,4}` | Matches **3 to 4** occurrences | `a{3,4}` → `aaa`, `aaaa` |

---

## 🧩 __6. Some commonly used functions are:__

| Function | Description |
|----------|-------------|
| `re.search()` | Searches for the first occurrence of a pattern. |
| `re.match()` | Matches a pattern only at the beginning of a string. |
| `re.fullmatch()` | Matches the entire string. |
| `re.findall()` | Returns all matches as a list. |
| `re.finditer()` | Returns an iterator over all matches. |
| `re.sub()` | Replaces matched patterns with another string. |
| `re.split()` | Splits a string based on a regex pattern. |
| `re.compile()` | Compiles a regex pattern for repeated use. |

---

> **Note:** Regex is powerful, but complex patterns can become difficult to read and maintain. Keep expressions as simple as possible.

````python
import re
````

### __1. re.findall()__

`# Characters & Anchors`

````python
pattern = r"\b\w{3}\b"  # match words with exactly 3 letters
text = "The cat sat on the mat."
result = re.findall(pattern, text)
print(result)  # Output: ['The', 'cat', 'sat', 'the', 'mat']
````

````python
pattern = r"\d+"  # match one or more digits
text = "There are 3 cats and 4 dogs."
result = re.findall(pattern, text)
print(result)  # Output: ['3', '4']
````

`# Quantifiers`

````python
# ? = 0 or 1
pattern = r"colou?r"  # match 'color' or 'colour'
text = "The color of the sky is blue, but some people spell it as colour and colouur."
result = re.findall(pattern, text)
print(result)  # Output: ['color', 'colour']
````

````python
# + = 1 or more
pattern = r"colou+r"  # match 'colour' with one or more 'u's
text = "The color of the sky is blue, but some people spell it as colour and colouur."
result = re.findall(pattern, text)
print(result)  # Output: ['colour', 'colouur']
````

````python
# * = 0 or more
pattern = r"colou*r"  # match 'color' or 'colour' with zero or more 'u's
text = "The color of the sky is blue, but some people spell it as colour and colouur."
result = re.findall(pattern, text)
print(result)  # Output: ['color', 'colour', 'colouur']
````

`Character Sets & Groups`

````python
pattern = r"cat|dog"  # match either 'cat' or 'dog'
text = "I have a cat and a dog."
result = re.findall(pattern, text)
print(result)  # Output: ['cat', 'dog']
````

````python
pattern = r"([a-zA-Z]+)@([a-zA-Z]+)\.([a-zA-Z]{2,})"  # match email addresses
text = "Please contact us at john@example.com or jane@test.org"
result = re.findall(pattern, text)
print(result)  # Output: [('john', 'example', 'com'), ('jane', 'test', 'org')]
````
