# __Word2Vec (Prediction-Based Word Embeddings)__

---

## __Why Do We Need Word2Vec?__

Earlier we learned **frequency-based text representations**:

- One-Hot Encoding
- Bag of Words (BoW)
- N-Grams
- TF-IDF

These methods represent words using **frequency/statistics**, not their meanings.

---

## 🚫 __Limitations of Frequency-Based Embeddings__

### __1. Semantic Meaning is Not Captured__

Words with similar meanings receive completely different vectors.

__Example:__

```
King ≠ Queen
Car ≠ Automobile
Dog ≠ Puppy
```

The model cannot understand that these words are related.

---

### __2. Sparsity Problem__

Most values in the vectors are **0**.

__Example (One-Hot Encoding):__

| Word | Vector |
|------|---------|
| Apple | [1,0,0,0,0] |
| Banana | [0,1,0,0,0] |
| Mango | [0,0,1,0,0] |

Most elements are zero → **Sparse Representation**

---

### __3. High Dimensional Vectors__

Vector size depends on vocabulary size.

If

Vocabulary = 50,000 words

then

```
One-Hot Vector Size = 50,000
```

As vocabulary grows, memory and computation increase significantly.

---

## ✅ __Prediction-Based Embeddings__

Instead of counting words,

**Word2Vec learns word representations by predicting surrounding words.**

```
Frequency Based
      ↓
Count words

Prediction Based
      ↓
Predict words
```

__Examples:__

- Word2Vec
- FastText
- GloVe (combines statistics + prediction ideas)

---

## __Intuition Behind Word2Vec__

Words that appear in **similar contexts** tend to have **similar meanings**.

This idea is known as:

> **"You shall know a word by the company it keeps."**
>
> — J. R. Firth

__Example:__

```
Apple launched a new iPhone.
```

Here,

```
Apple → Company
```

because nearby words are

```
launched
new
iPhone
```

---

__Another example:__

```
I like to eat an apple.
```

Here,

```
Apple → Fruit
```

because nearby words are

```
eat
like
fruit
```

Word2Vec learns meaning from **context**, not from dictionaries.

---

## __Neural Network Intuition__

Word2Vec is built using a simple Neural Network.

Let's first understand how a neural network learns.

### __Example:__ 🏡 House Price Prediction

| Area (sq.ft.) | Location | Price |
|---------------|----------|--------|
| 500 | City | ₹5 Lakh |
| 800 | City | ₹8 Lakh |
| 1000 | Village | ? |

The price depends on multiple features.

```
X₁ = Area

X₂ = Location

Y = Price
```

Initially,

```
Y = WX
```

For multiple inputs,

```
Y = W₁X₁ + W₂X₂
```

where

- W₁ = Importance of Area
- W₂ = Importance of Location

During training, the neural network learns these weights automatically.

---

## __Neural Networks are Universal Function Approximators__

A neural network can learn **complex relationships** between inputs and outputs by adjusting its weights during training.

This learning capability is what enables Word2Vec to discover semantic relationships between words.

---

## __How Word2Vec Captures Semantic Meaning__

Imagine every word is represented by several hidden features.

| Feature | King | Queen | Man | Woman |
|---------|------|--------|-----|-------|
| Power | 0.9 | 0.8 | 0.4 | 0.4 |
| Royalty | 1.0 | 1.0 | 0.2 | 0.2 |
| Masculinity | 0.9 | 0.1 | 0.9 | 0.1 |
| Femininity | 0.1 | 0.9 | 0.1 | 0.9 |
| Can You Eat? | 0 | 0 | 0 | 0 |

These are **not manually created features**.

Word2Vec automatically learns them during training.

---

## __Word Vectors__

```
King
=
[0.9, 1, 0.9, 0.1, 0]

Queen
=
[0.8, 1, 0.1, 0.9, 0]
```

---

## __Word Vector Arithmetic__

One fascinating property of Word2Vec is that vector arithmetic captures semantic relationships.

__Example:__

```
King − Man + Woman
```

Let's compute each feature.

| Feature | Calculation | Result |
|----------|-------------|---------|
| Power | 0.9 − 0.4 + 0.4 | 0.9 |
| Royalty | 1 − 0.2 + 0.2 | 1 |
| Masculinity | 0.9 − 0.9 + 0.1 | 0.1 |
| Femininity | 0.1 − 0.1 + 0.9 | 0.9 |
| Can You Eat? | 0 − 0 + 0 | 0 |

Resulting vector:

```
[0.9, 1, 0.1, 0.9, 0]
```

Compare it with

```
Queen

[0.8, 1, 0.1, 0.9, 0]
```

They are very close.

Hence,

```
King - Man + Woman ≈ Queen
```

---

## __Why Does This Work?__

Word2Vec places words with similar meanings close together in a high-dimensional vector space.

```
King -------- Queen

Man -------- Woman

Prince ----- Princess

Paris ------ France

Tokyo ------ Japan
```

Semantic relationships become mathematical relationships.

---

## __Key Takeaways__

- Word2Vec is a **prediction-based embedding technique**.
- It learns word meanings from surrounding context.
- Similar words have similar vectors.
- Dense vectors solve the sparsity problem.
- Vector size is fixed (typically 100–300 dimensions), regardless of vocabulary size.
- Word vectors capture semantic relationships.
- Vector arithmetic can reveal analogies such as:

```
King - Man + Woman ≈ Queen
```

# 📏 __Similarity Metrics__

Word embeddings are represented as vectors in a high-dimensional space.

To determine how similar two words are, we compute the **distance** or **similarity** between their vectors.

---

## __1. Manhattan Distance (L1 Distance)__

### __Formula__

```text
Distance = |x₁ - x₂| + |y₁ - y₂|
```

### __Interpretation__

- ✅ Smaller distance → More similar
- ❌ Larger distance → Less similar

---

## __2. Euclidean Distance (L2 Distance)__

### __Formula__

```text
            ______________________________
Distance = √((x₁ - x₂)² + (y₁ - y₂)²)
```

### __Interpretation__

- ✅ Smaller distance → More similar
- ❌ Larger distance → Less similar

---

## __3. Cosine Similarity ⭐ (Most Common)__

### _Formula_

```text
Cosine Similarity = cos(θ)
```

or

```text
               A · B
cos(θ) = -------------------
          ||A|| × ||B||
```

### __Range__

```
-1  ----------------  0  ----------------  1

Opposite           Orthogonal          Identical Direction
```

### __Interpretation__

- ✅ Closer to **1** → Highly similar
- ⚪ Around **0** → Unrelated
- ❌ Closer to **-1** → Opposite direction

---

## __4. Dot Product__

### __Formula__

```text
Dot Product = ||A|| × ||B|| × cos(θ)
```

### __Interpretation__

- ✅ Larger value → More similar
- ❌ Smaller value → Less similar

> Unlike cosine similarity, the dot product is affected by the magnitude (length) of the vectors.

---

# __Why Do We Prefer Cosine Similarity?__

### __Advantages__

- Ignores vector magnitude and focuses only on direction.
- More aligned with human intuition for measuring semantic similarity.
- Fixed range **[-1, 1]**, making it easy to interpret.
- Widely used in NLP applications such as Word2Vec, GloVe, BERT, and semantic search.

---

# __Quick Comparison__

| Metric | Measures | Better Similarity |
|---------|----------|-------------------|
| Manhattan Distance | Absolute distance | Smaller value |
| Euclidean Distance | Straight-line distance | Smaller value |
| Cosine Similarity ⭐ | Angle between vectors | Larger value (Closer to 1) |
| Dot Product | Angle + Magnitude | Larger value |

---

# __Key Takeaway__

> **Distance Metrics**
>
> - Manhattan Distance
> - Euclidean Distance
>
> **Similarity Metrics**
>
> - Cosine Similarity ⭐ (Most Preferred)
> - Dot Product

````python
# ============================================
# Download Dataset from Google Drive
# ============================================

import gdown

# Direct download URL of the dataset hosted on Google Drive
url = "https://drive.google.com/uc?export=download&id=1RvvQlv0U9SpD_KEU1DZnL13iRuKcu9-L"

# Local filename to save the downloaded dataset
output = "text.txt"

# Download the dataset
# quiet=False displays the download progress
gdown.download(url, output, quiet=False)


# ============================================
# Read the Downloaded Dataset
# ============================================

# Open the downloaded text file in read mode using UTF-8 encoding
with open(output, "r", encoding="utf-8") as file:
    text = file.read()

# Preview the first 150 characters (optional)
text[:150]
````

````python
# ============================================
# Text Preprocessing
# ============================================

import re

# Keep only:
# - Uppercase and lowercase English letters (A-Z, a-z)
# - Spaces
# - Periods (.) to preserve sentence boundaries
#
# Remove:
# - Numbers
# - Special characters
# - Punctuation other than periods

pattern = r"[^a-zA-Z. ]"
preprocessed_text = re.sub(pattern, "", text)

# Preview the first 150 characters
preprocessed_text[:150]
````

````python
# ============================================
# Convert Text into Sentences
# ============================================

# Split the cleaned text into individual sentences
# using the period (.) as the sentence delimiter.

sentences = preprocessed_text.split(".")

# Display the first few sentences
sentences[:5]
````

````python
# ============================================
# Tokenize Sentences
# ============================================

# Convert each sentence into a list of words.
#
# Example:
# "Harry was happy"
#      ↓
# ["Harry", "was", "happy"]

sentence_list = [sentence.split() for sentence in sentences]

# Display the first two tokenized sentences
sentence_list[:2]
````

````python
# ============================================
# Train the Word2Vec Model
# ============================================

from gensim.models import Word2Vec

model = Word2Vec(
    sentences=sentence_list,   # Tokenized training corpus
    vector_size=100,           # Dimension of each word embedding
    window=3,                  # Context window size
    sg=0                       # 0 = CBOW | 1 = Skip-Gram
)

print("Word2Vec model trained successfully!")
````

## 🧪 __Testing the Trained Model__

After training, we can explore the learned word embeddings and verify whether the model has captured semantic relationships between words.

````python
# ============================================
# Access Word Embeddings
# ============================================

# Retrieve the 100-dimensional vector representation
# for the word "Harry".

model.wv["Harry"]
````

````python
# ============================================
# Check Embedding Dimension
# ============================================

# Verify the size of the generated word vector.

len(model.wv["Harry"])
````

````python
# ============================================
# Find Similar Words
# ============================================

# Retrieve words that are most similar to "Hermione"
# based on cosine similarity.

model.wv.most_similar("Hermione")
````

````python
# ============================================
# Testing Unknown Words
# ============================================

# Check how the model behaves when querying
# a word that does not exist in the training corpus.

model.wv.most_similar("apple")
````

# 📖 __Word2Vec Training Methods__

Word2Vec provides two different architectures for learning word embeddings.

## __1. Continuous Bag of Words (CBOW)__

### __Idea__

CBOW predicts the **current (target) word** using the surrounding context words.

__Example:__

```text
Sentence:
"The cat sat on the mat"

Context:
"The cat ___ on the mat"

Target:
sat
```

The model learns to predict **"sat"** from its neighboring words.

### __Characteristics__

- Faster to train
- Works well for larger datasets
- Better for frequent words
- Less computationally expensive

### __Visualization__

```text
Context Words
(The, cat, on, the, mat)
          │
          ▼
     Neural Network
          │
          ▼
     Predict Target
          "sat"
```

---

## __2. Skip-Gram__

### __Idea__

Skip-Gram does the opposite of CBOW.

Given the **current word**, it predicts the surrounding context words.

__Example:__

```text
Target:
sat

Predicted Context:
The
cat
on
the
mat
```

### __Characteristics__

- Slower to train
- Performs better on smaller datasets
- Learns better representations for rare words
- Generally produces higher-quality embeddings

### __Visualization__

```text
Target Word
    "sat"
      │
      ▼
Neural Network
      │
      ▼
Predict Context Words
(The, cat, on, the, mat)
```

---

# 🔄 __CBOW vs Skip-Gram__

| Feature | CBOW | Skip-Gram |
|---------|------|-----------|
| Predicts | Target word | Context words |
| Input | Context | Target |
| Speed | Faster | Slower |
| Accuracy | Good | Better |
| Rare Words | Not as effective | Better |
| Large Dataset | Excellent | Good |
| Small Dataset | Good | Excellent |

---

# ⚙️ __Important Hyperparameters__

| Parameter | Description |
|------------|-------------|
| `vector_size` | Size of each word embedding |
| `window` | Number of neighboring words considered as context |
| `min_count` | Ignore words appearing fewer than this number |
| `workers` | Number of CPU cores used during training |
| `sg` | Training algorithm (`0 = CBOW`, `1 = Skip-Gram`) |
| `epochs` | Number of training iterations |
| `negative` | Number of negative samples used during training |

__Example:__

```python
from gensim.models import Word2Vec

model = Word2Vec(
    sentences=tokenized_sentences,
    vector_size=100,
    window=5,
    min_count=2,
    workers=4,
    sg=1,          # Skip-Gram
    epochs=10
)
```
